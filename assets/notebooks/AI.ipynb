## 1.1 What is Artificial Intelligence?

Artificial Intelligence (AI) is the field of computer science dedicated to creating systems that **exhibit intelligent behaviour**, reasoning, learning, problem-solving, perception, and language understanding.

### Two Paradigms of AI

- Uses explicit rules, logic, and knowledge representations

- Strengths: Interpretable, deterministic
- Weaknesses: Brittle, hard to scale

**2. Computational Intelligence / Bio-inspired AI**:
- Learns patterns from data or imitates nature
- Examples: Neural networks, evolutionary algorithms, swarm intelligence
- Strengths: Flexible, robust, scalable
- Weaknesses: Often a black box

This course focuses on **meta-heuristics**, a class of algorithms that sit at the intersection of AI, operations research, and biology. They are:

- **General-purpose**: Work across many problem types
- **Approximate**: Find *good* solutions, not necessarily *optimal*
- **Efficient**: Scale to large, real-world problems
- **Inspired by nature**: Evolution, ant colonies, thermodynamics

## 1.2 Optimization Problems: The Core Challenge

### Formal Definition

An **optimization problem** is defined by a triple:

$$\mathcal{P} = (S, f, \Omega)$$

Where:
- $S$ = **Search space** (all possible candidate solutions)
- $f: S \rightarrow \mathbb{R}$ = **Objective function** (measures solution quality)
- $\Omega$ = **Set of constraints** (defines feasibility)

**Goal**: Find $s^* \in S$ such that:

$$f(s^*) \leq f(s) \quad \forall s \in S \quad \text{(minimization)}$$

or equivalently:

$$f(s^*) \geq f(s) \quad \forall s \in S \quad \text{(maximization)}$$

### Types of Optimization Problems

#### Continuous Optimization

The search space is a continuous domain, e.g., $S \subseteq \mathbb{R}^n$.

- **Example**: Minimise $f(x_1, x_2) = x_1^2 + x_2^2$ over $\mathbb{R}^2$
- Classical methods: Gradient Descent, Newton's method
- These work when $f$ is **differentiable and convex**

The gradient descent update rule:

$$\mathbf{x}^{(t+1)} = \mathbf{x}^{(t)} - \eta \nabla f(\mathbf{x}^{(t)})$$

where $\eta > 0$ is the **learning rate** and $\nabla f$ is the gradient:

$$\nabla f = \left( \frac{\partial f}{\partial x_1}, \frac{\partial f}{\partial x_2}, \ldots, \frac{\partial f}{\partial x_n} \right)$$

#### Combinatorial Optimization

The search space is **discrete** (finite or countably infinite).

- **Example**: Travelling Salesman Problem (TSP), Knapsack, Vehicle Routing
- Classical methods: Dynamic Programming, Branch & Bound
- These quickly become **intractable** for large instances

### Complexity: Why Optimization Is Hard

**P vs NP**: The most famous open question in computer science.

- **P**: Problems solvable in polynomial time $O(n^k)$
- **NP**: Problems *verifiable* in polynomial time
- **NP-hard**: At least as hard as any NP problem
- **NP-complete**: NP-hard AND in NP

Most real-world combinatorial optimization problems are **NP-hard**:

| Problem | Complexity | Search Space Size |
|---------|------------|-------------------|
| TSP (n cities) | NP-hard | $(n-1)!/2$ tours |
| Knapsack | NP-hard | $2^n$ subsets |
| VRP | NP-hard | Larger than TSP |
| Graph Coloring | NP-complete | Exponential |

For TSP with $n=20$ cities: $(20-1)!/2 \approx 6 \times 10^{16}$ possible tours!

For TSP with $n=100$: $(99)!/2 \approx 4.7 \times 10^{155}$, more than atoms in the universe.

**This is why meta-heuristics exist**: they sacrifice optimality guarantees for practicality.

## 1.3 Search Spaces and Fitness Landscapes

### The Fitness Landscape Metaphor

Imagine the search space as a **geographical landscape**:
- Each point on the landscape = a candidate solution
- Altitude = objective function value
- We want to find the **highest peak** (maximization) or **lowest valley** (minimization)

### Mathematical Definition of a Fitness Landscape

A fitness landscape is a tuple:

$$\mathcal{L} = (S, f, d)$$

Where:
- $S$ = solution space
- $f: S \rightarrow \mathbb{R}$ = fitness function
- $d: S \times S \rightarrow \mathbb{R}^{\geq 0}$ = distance metric

### Types of Landscapes

#### 1. Unimodal Landscape
- One global optimum, no local optima
- Easy for gradient-based methods
- Example: $f(x) = x^2$ (one minimum at $x=0$)

#### 2. Multimodal Landscape
- Multiple local optima
- Gradient descent gets **trapped in local optima**
- Example: $f(x) = x\sin(x)$ has many local minima

The Rastrigin function is a classic test case:

$$f(\mathbf{x}) = An + \sum_{i=1}^{n} \left[ x_i^2 - A\cos(2\pi x_i) \right]$$

with $A = 10$. This has approximately $10^n$ local minima, a nightmare for gradient methods!

#### 3. Rugged Landscape
- Many local optima of similar quality
- Typical of combinatorial problems (TSP, VRP)
- The "distance" between solutions is defined by neighbourhood structure

#### 4. Deceptive Landscape
- The path of best fitness does NOT lead to the global optimum
- Most challenging for evolutionary algorithms
- Example: A plateau that looks like it leads somewhere but doesn't

### Local vs Global Optima

**Local Optimum**: A solution $s^*_{local}$ such that:
$$f(s^*_{local}) \leq f(s) \quad \forall s \in \mathcal{N}(s^*_{local})$$

where $\mathcal{N}(s)$ is the **neighbourhood** of solution $s$.

**Global Optimum**: A solution $s^*$ such that:
$$f(s^*) \leq f(s) \quad \forall s \in S$$

Every global optimum is a local optimum, but **not vice versa**.

### The Exploration–Exploitation Trade-off

This is the **central dilemma** of all meta-heuristics:

| | Exploration | Exploitation |
|-|-------------|---------------|
| **What** | Search new, unexplored regions | Refine solutions in known good regions |
| **Benefit** | Avoids local optima | Converges to high-quality solutions |
| **Risk** | Wastes time on bad regions | Premature convergence |

**Too much exploration** → Random walk, never converges
**Too much exploitation** → Gets stuck in local optima

Every meta-heuristic we study has a mechanism to balance this trade-off.

## 1.4 Why Classical Methods Fail: The Curse of Dimensionality

### The Curse of Dimensionality (Bellman, 1957)

As the number of dimensions $n$ grows, the volume of the search space grows **exponentially**.

**Example**: Suppose each variable $x_i \in [0, 1]$ is discretized into $k$ values.
- For $n=1$: $k$ candidate solutions
- For $n=2$: $k^2$ candidate solutions
- For $n=10$: $k^{10}$ candidate solutions
- For $n=100$: $k^{100}$ candidate solutions

With $k=10$ and $n=100$: $10^{100}$ candidates, a **googol** of solutions!

No computer in the universe can enumerate all of these.

### Why Gradient Methods Fail on Real Problems

Classical gradient-based methods require:
1. **Differentiability**: Many real-world objectives are not differentiable (e.g., tour length in TSP)
2. **Convexity**: Most real problems are non-convex (multiple local optima)
3. **Continuity**: Combinatorial problems have discrete spaces
4. **Known structure**: Many problems are black boxes

### Why Exact Methods Fail

**Branch and Bound** for TSP:
- Provably finds the optimal solution
- But runtime is $O(2^n)$ in the worst case
- Solves TSP up to ~100 cities practically
- Fails for VRP with 1000+ customers

**Dynamic Programming** for Knapsack:
- Runtime: $O(nW)$ where $W$ = capacity
- This is **pseudo-polynomial**, exponential in the binary encoding of $W$

### The Practical Need for Approximation

Real-world problems need solutions **now**, not in a million years:
- A delivery company must plan 1000 routes in seconds
- A factory must schedule 500 jobs by morning
- A telecom must assign frequencies to 10,000 towers today

The question is no longer *"Is this optimal?"* but:

> *"How good can we get in the time we have?"*

This is the practical motivation for **all meta-heuristics**.

## 1.5 The Taxonomy of Meta-Heuristics

A **meta-heuristic** is a high-level strategy (a "strategy about strategies") that guides an underlying search procedure.

The word comes from Greek:
- *meta* = beyond, higher-level
- *heuristikós* = "good at discovering" (from *heuriskein* = to find)

### Classification 1: Single-solution vs Population-based

```
META-HEURISTICS
│
├── Single-Solution (trajectory methods)
│   ├── Simulated Annealing (SA)
│   ├── Tabu Search (TS)
│   └── Iterated Local Search (ILS)
│
└── Population-Based
    ├── Evolutionary Algorithms
    │   ├── Genetic Algorithms (GA)
    │   ├── Evolution Strategies (ES)
    │   └── PBIL
    └── Swarm Intelligence
        ├── Ant Colony Optimization (ACO)
        └── Particle Swarm Optimization (PSO)
```

### Classification 2: Nature of Inspiration

| Inspiration | Algorithm | Analogy |
|-------------|-----------|----------|
| Biology / Evolution | Genetic Algorithms | Natural selection |
| Ant Behaviour | ACO | Pheromone trails |
| Physics (annealing) | Simulated Annealing | Cooling metal |
| Memory | Tabu Search | Short-term memory |
| Bird flocking | PSO | Swarm navigation |
| Perturbation | ILS | Kick & improve |

### Key Shared Properties

All meta-heuristics share these design principles:

1. **Stochastic**: They use randomness to explore the search space
2. **Approximate**: They do not guarantee finding the global optimum
3. **General**: They can be applied to many problem types
4. **Black-box compatible**: They only need to evaluate $f(s)$, not its structure
5. **Iterative**: They improve solutions over many iterations

## 1.6 The No Free Lunch Theorem

### Statement of the Theorem

The **No Free Lunch (NFL) Theorem** (Wolpert & Macready, 1997) states:

> *If algorithm A outperforms algorithm B on a set of problems $P$, then algorithm B will outperform algorithm A on the complementary set of problems $\bar{P}$.*

More formally, for any two optimization algorithms $a_1$ and $a_2$:

$$\sum_{f} P(d_y^m | f, m, a_1) = \sum_{f} P(d_y^m | f, m, a_2)$$

Where:
- $f$ = objective function (summed over **all possible** objective functions)
- $m$ = number of function evaluations
- $d_y^m$ = sequence of objective values observed
- $P(d_y^m | f, m, a)$ = probability of observing $d_y^m$ given function $f$, $m$ evaluations, and algorithm $a$

### What Does This Mean?

**Averaged over ALL possible problems**, every algorithm performs equally well.

**Implications**:
- There is NO universally best optimization algorithm
- An algorithm's superiority always comes from **problem-specific knowledge**
- We must **match algorithms to problem structure**

### Why Meta-Heuristics Work Despite NFL

NFL is a result about **all possible** functions, including pathological ones that never occur in practice.

In the real world:
- Problems have **structure** (nearby solutions tend to have similar fitness)
- This property is called **locality** or **smoothness**
- Meta-heuristics are designed to exploit this structure

**Example**: In TSP, two tours that share 99 edges out of 100 usually have similar lengths. ACO and GA exploit this by building solutions using good partial solutions.

### Practical Lesson

When choosing an algorithm, ask:
1. What is the structure of my search space?
2. What is the nature of my objective function?
3. What neighbourhood relationships exist between solutions?
4. Do I need exact optima or good approximations?

### Q1: What is the difference between a heuristic and a meta-heuristic?

**Answer**:

A **heuristic** is a problem-specific rule of thumb. For example, in TSP, the *nearest-neighbour heuristic* says: "Always go to the closest unvisited city." It works reasonably well for TSP but has no general applicability.

A **meta-heuristic** is a *higher-level strategy* that is **problem-independent**. It defines a general framework (e.g., "maintain a memory of previously visited solutions and avoid revisiting them") that can be instantiated for any specific problem by plugging in a problem-specific representation and objective function.

Think of it this way:
- Heuristic = a recipe for one dish
- Meta-heuristic = a general cooking philosophy (e.g., French cuisine) that guides creating many recipes

---

### Q2: What does it mean for a problem to be NP-hard, and why does it matter for choosing algorithms?

**Answer**:

A problem is **NP-hard** if every problem in NP can be reduced to it in polynomial time. Informally, it means we believe no algorithm can solve all instances of the problem in polynomial time (assuming P ≠ NP).

For optimization, this means:
- No exact algorithm is known to run efficiently for large instances
- As problem size grows, runtime explodes exponentially
- For real-world-scale instances (100s–1000s of variables), exact methods are infeasible

This motivates meta-heuristics: they give up the **optimality guarantee** in exchange for **practical efficiency**. They find *good-enough* solutions in *reasonable time*.

### Q3: Explain the exploration–exploitation trade-off and give an example from everyday life.

**Answer**:

The **exploration–exploitation dilemma** arises in any sequential decision-making process:
- **Exploration**: Try new, untested options to discover potentially better ones
- **Exploitation**: Use the best option you already know about

**Everyday example**: Choosing a restaurant for dinner.
- **Exploitation**: Go to your favourite restaurant (you know it's good, but you might miss better places)
- **Exploration**: Try a new place (you might find somewhere amazing, or waste an evening)

In optimization algorithms:
- Pure exploitation → converges to local optima
- Pure exploration → random walk, never improves
- Good balance → finds global optimum reliably

Each meta-heuristic has a different mechanism for achieving this balance:
- SA: temperature controls randomness (high temp = exploration, low temp = exploitation)
- ACO: pheromones reinforce good solutions (exploitation) while randomness promotes diversity (exploration)
- GA: crossover combines known good solutions (exploitation), mutation introduces novelty (exploration)

### Q4: The No Free Lunch theorem says all algorithms are equal. Does this mean we should just pick any algorithm randomly?

**Answer**:

**No!** NFL averages over ALL possible functions, including completely random and adversarially constructed ones that never appear in practice.

Real optimization problems are NOT random, they have **structure**:
- Similar inputs tend to give similar outputs (locality)
- Good partial solutions can often be combined into good complete solutions (composability)
- Problem-specific knowledge can guide search (exploitable regularities)

The lesson of NFL is: **use problem knowledge to choose your algorithm**. An algorithm that exploits the specific structure of your problem will vastly outperform one that ignores it.

AFL says: "for free lunches, everyone gets the same meal." But if you know the menu and order strategically, you eat much better.

---

### Q5: Prove that the number of tours in TSP with $n$ cities is $(n-1)!/2$.

**Answer**:

**Proof**:

1. We have $n$ cities. A tour visits all $n$ cities exactly once and returns to the start.
2. Fix the starting city (WLOG, city 1). The remaining $n-1$ cities can be arranged in $(n-1)!$ orderings.
3. Each tour can be traversed in **two directions** (clockwise and counterclockwise), and both represent the same tour (same edges, same cost). So we divide by 2.

Therefore, the number of distinct tours = $\dfrac{(n-1)!}{2}$.

**Verification**: For $n=4$ cities: $3!/2 = 3$ distinct tours. We can verify manually:
- Cities {1,2,3,4}: Tours are 1→2→3→4→1, 1→2→4→3→1, 1→3→2→4→1. ✓

## 2.1 Biological Inspiration: Real Ants

### The Double Bridge Experiment (Deneubourg et al., 1990)

A classic biology experiment with Argentine ants (*Iridomyrmex humilis*):

**Setup**: Two paths of *different lengths* connect a nest to a food source.
- Path A: Short path
- Path B: Long path

**What happens**:
1. Initially, ants choose paths randomly (50/50 split)
2. Ants on the **shorter path** return faster, depositing pheromone more frequently per unit time
3. More pheromone on the short path → more ants choose it → even more pheromone → ...
4. Within 30–60 minutes, **almost all ants converge on the shortest path**

This is **positive feedback through stigmergy**:
- *Stigmergy*: indirect coordination through environmental modification
- No ant "knows" about the other path, the colony intelligence emerges from local interactions

### Key Properties of Ant Behaviour

1. **Pheromone deposition**: Ants deposit pheromone on their path
2. **Pheromone following**: Ants prefer paths with more pheromone
3. **Pheromone evaporation**: Pheromone decays over time (prevents permanent commitment to suboptimal paths)
4. **Probabilistic choice**: Ants do not deterministically follow the best path, they maintain diversity

### From Biology to Algorithm

| Biology | ACO Algorithm |
|---------|---------------|
| Ant | Artificial ant (algorithm agent) |
| Path in environment | Edge in graph |
| Pheromone trail | Pheromone value $\tau_{ij}$ on edge $(i,j)$ |
| Path quality = shorter → more pheromone | Better solution → more pheromone |
| Evaporation | $\tau_{ij} \leftarrow (1-\rho)\tau_{ij}$ |
| Probabilistic path choice | Probability formula using pheromone + heuristic |

## 2.2 The ACO Framework

### Problem Representation as a Graph

ACO operates on a **construction graph** $G = (C, L)$ where:
- $C = \{c_1, c_2, \ldots, c_{|C|}\}$ = set of **components** (solution building blocks)
- $L \subseteq C \times C$ = set of **connections** between components
- $\tau_{ij}$ = pheromone value on edge $(c_i, c_j)$
- $\eta_{ij}$ = heuristic value on edge $(c_i, c_j)$ (problem-specific desirability)

**For TSP**: Cities are nodes, edges connect all pairs of cities, $\eta_{ij} = 1/d_{ij}$ (inverse distance, shorter is more desirable).

### The ACO Meta-Heuristic (Dorigo & Stützle, 2004)

The general ACO meta-heuristic runs as follows:

```
PROCEDURE ACO_Meta_Heuristic:
    Initialise pheromone trails τ
    WHILE (termination criterion not met) DO:
        ConstructAntSolutions()
        ApplyLocalSearch()           [optional but important!]
        UpdatePheromones()
    END WHILE
    RETURN best solution found
```

### Three Core Components of Every ACO Algorithm

**1. Solution Construction**:
Each ant builds a solution by moving through the construction graph, choosing components probabilistically based on pheromone ($\tau$) and heuristic ($\eta$) information.

**2. Pheromone Update**:
After all ants construct solutions, pheromone values are updated:
- **Evaporation**: All trails decay (forget bad solutions, prevent stagnation)
- **Deposition**: Good solutions deposit more pheromone (reinforce good choices)

**3. Local Search** (optional but highly effective):
Before updating pheromones, apply local search to improve ant solutions.

## 2.3 Pheromone Models and Stigmergy

### Pheromone Initialisation

All pheromone values are initialised to a small positive constant:

$$\tau_{ij}(0) = \tau_0 > 0 \quad \forall (i,j) \in L$$

Why not zero? To ensure every edge has a non-zero probability of being chosen initially (promotes exploration).

### Pheromone Evaporation

After each iteration, all pheromone trails are reduced:

$$\tau_{ij} \leftarrow (1 - \rho) \cdot \tau_{ij}$$

where $\rho \in (0, 1]$ is the **evaporation rate**.

**Purpose**: Without evaporation, pheromone would accumulate forever, locking the algorithm into suboptimal solutions. Evaporation implements **forgetting**, it allows the algorithm to explore new regions.

**Mathematical effect**: After $t$ iterations of evaporation with no deposition:

$$\tau_{ij}(t) = (1-\rho)^t \cdot \tau_0 \rightarrow 0 \quad \text{as } t \rightarrow \infty$$

### Pheromone Deposition

Ants that found good solutions deposit pheromone on the edges they used:

$$\tau_{ij} \leftarrow \tau_{ij} + \sum_{k=1}^{m} \Delta\tau_{ij}^k$$

Where $m$ is the number of ants, and $\Delta\tau_{ij}^k$ is the amount of pheromone deposited by ant $k$ on edge $(i,j)$:

$$\Delta\tau_{ij}^k = \begin{cases} Q / L_k & \text{if ant } k \text{ used edge } (i,j) \\ 0 & \text{otherwise} \end{cases}$$

Here:
- $Q$ = a positive constant
- $L_k$ = length (cost) of the tour found by ant $k$
- **Better tours (smaller $L_k$) deposit more pheromone** → reinforcement learning!

### The Role of $\rho$ (Evaporation Rate)

| $\rho$ value | Effect |
|-------------|--------|
| High (close to 1) | Fast forgetting → high exploration, slow convergence |
| Low (close to 0) | Slow forgetting → strong exploitation, risk of stagnation |
| Typical range | $0.1$ to $0.5$ |

## 2.4 The Travelling Salesman Problem (TSP)

The **Travelling Salesman Problem** is the most famous combinatorial optimization problem and the primary benchmark for ACO.

### Problem Definition

**Given**: A set of $n$ cities with coordinates, and a distance matrix $D$ where $d_{ij}$ = distance between cities $i$ and $j$.

**Find**: A Hamiltonian cycle (tour) that visits every city exactly once and returns to the start, with **minimum total length**.

$$\text{Minimise} \quad L(\pi) = \sum_{i=1}^{n} d_{\pi(i), \pi(i+1)}$$

where $\pi = (\pi(1), \pi(2), \ldots, \pi(n))$ is a permutation of cities and $\pi(n+1) = \pi(1)$.

### Variants of TSP

| Variant | Description |
|---------|-------------|
| Symmetric TSP (sTSP) | $d_{ij} = d_{ji}$ (undirected) |
| Asymmetric TSP (aTSP) | $d_{ij} \neq d_{ji}$ (directed, e.g., one-way streets) |
| Euclidean TSP | Cities in 2D plane, $d_{ij} = \sqrt{(x_i-x_j)^2 + (y_i-y_j)^2}$ |
| Multi-depot TSP | Multiple starting points |

### Why TSP Is Hard

The number of possible tours grows **super-exponentially** with $n$:

| $n$ | Number of tours | Time to check all (at $10^9$/sec) |
|-----|-----------------|-----------------------------------|
| 10 | 181,440 | < 1 second |
| 20 | $6 \times 10^{16}$ | ~2 years |
| 50 | $1.5 \times 10^{62}$ | $10^{45}$ years |
| 100 | $4.7 \times 10^{155}$ | Unfathomable |

The nearest-neighbour heuristic gives solutions typically 20–25% above optimal. Good meta-heuristics (ACS + local search) reach within **0.1–1% of optimal** for large instances.

### TSP as a Construction Graph

For ACO, TSP is represented as:
- **Nodes**: $n$ cities
- **Edges**: All $\binom{n}{2}$ pairs (complete graph)
- **Pheromone**: $\tau_{ij}$ on edge $(i,j)$ = desirability of visiting $j$ after $i$
- **Heuristic**: $\eta_{ij} = 1/d_{ij}$ (prefer closer cities)

## 2.5 Ant System (AS): The First ACO Algorithm

**Ant System** was proposed by Colorni, Dorigo, and Maniezzo (1991, 1992), the first ACO algorithm.

### The Transition Probability Rule

At each step, ant $k$ currently at city $i$ chooses the next city $j$ to visit with probability:

$$p_{ij}^k = \frac{[\tau_{ij}]^\alpha \cdot [\eta_{ij}]^\beta}{\sum_{l \in J_i^k} [\tau_{il}]^\alpha \cdot [\eta_{il}]^\beta}$$

Where:
- $\tau_{ij}$ = pheromone on edge $(i,j)$
- $\eta_{ij} = 1/d_{ij}$ = heuristic desirability
- $\alpha \geq 0$ = controls influence of pheromone (exploitation)
- $\beta \geq 0$ = controls influence of heuristic (greediness)
- $J_i^k$ = set of cities not yet visited by ant $k$ (tabu list)

**Understanding the formula**:
- When $\alpha = 0$: ignore pheromone → greedy nearest-neighbour
- When $\beta = 0$: ignore heuristic → pure pheromone guidance
- Typical values: $\alpha \in [1, 2]$, $\beta \in [2, 5]$

### The AS Pheromone Update

After all $m$ ants complete their tours:

**Step 1: Evaporation**
$$\tau_{ij} \leftarrow (1 - \rho) \cdot \tau_{ij}, \quad \forall (i,j)$$

**Step 2: Deposition**
$$\tau_{ij} \leftarrow \tau_{ij} + \sum_{k=1}^{m} \Delta\tau_{ij}^k$$

With the deposition formula (AS-quantity variant):

$$\Delta\tau_{ij}^k = \begin{cases} \frac{Q}{L_k} & \text{if ant } k \text{ used edge } (i,j) \\ 0 & \text{otherwise} \end{cases}$$

### Understanding the Reinforcement

Why divide by $L_k$? Because:
- Short tour ($L_k$ small) → large $\Delta\tau$ → more reinforcement
- Long tour ($L_k$ large) → small $\Delta\tau$ → less reinforcement
- This implements the principle: **reward good solutions more strongly**

### Complete AS Algorithm

```
ALGORITHM Ant_System(n, m, α, β, ρ, Q, max_iter):
    INPUT: n cities, m ants, parameters α, β, ρ, Q
    
    // Initialisation
    τ_ij ← τ_0 = m / L_nn    for all (i,j)
    // L_nn = nearest-neighbour tour length (good heuristic estimate)
    
    best_tour ← ∅
    best_length ← ∞
    
    FOR iter = 1 TO max_iter:
        // Construct solutions
        FOR ant k = 1 TO m:
            Place ant k at random starting city
            WHILE tour not complete:
                Choose next city j with probability p_ij^k
                Move ant to city j; add j to tabu list
            Complete tour (return to start)
            Compute L_k (tour length)
            IF L_k < best_length:
                best_tour ← current tour
                best_length ← L_k
        
        // Update pheromone
        FOR all edges (i,j):
            τ_ij ← (1-ρ) · τ_ij
            FOR ant k = 1 TO m:
                τ_ij ← τ_ij + Δτ_ij^k
    
    RETURN best_tour, best_length
```

### Limitations of AS

AS was a breakthrough but had weaknesses:
1. **Slow convergence**: All ants update pheromone equally
2. **All ants are equal**: Even very bad tours deposit pheromone
3. **No exploitation of best-so-far**: Doesn't leverage the globally best solution found
4. **Stagnation**: Can converge prematurely to suboptimal solutions

These limitations motivated the development of **Ant Colony System (ACS)**.

## 2.6 Ant Colony System (ACS): The Landmark Algorithm

**ACS** was developed by **Dorigo and Gambardella (1997)**, a landmark paper that dramatically improved performance over AS. This is a central algorithm in Prof. Gambardella's research.

### Three Key Innovations of ACS Over AS

| Feature | Ant System (AS) | Ant Colony System (ACS) |
|---------|-----------------|-------------------------|
| Transition rule | Proportional (always probabilistic) | Pseudo-random proportional |
| Pheromone update | All ants update | Only best ant updates |
| Local update | None | Online pheromone update during construction |

### Innovation 1: Pseudo-Random Proportional Rule

ACS uses a **two-phase decision rule** with a random variable $q \sim U[0,1]$ and parameter $q_0 \in [0,1]$:

$$j = \begin{cases} \arg\max_{l \in J_i^k} \left[ \tau_{il} \cdot [\eta_{il}]^\beta \right] & \text{if } q \leq q_0 \quad \text{(exploitation)} \\ J \sim p_{ij}^k & \text{if } q > q_0 \quad \text{(exploration)} \end{cases}$$

Where during exploration, the probability is:

$$p_{ij}^k = \frac{\tau_{ij} \cdot [\eta_{ij}]^\beta}{\sum_{l \in J_i^k} \tau_{il} \cdot [\eta_{il}]^\beta}$$

**Key insight**: When $q \leq q_0$, the ant greedily picks the **best-known** edge (exploitation). When $q > q_0$, it uses randomness (exploration). The parameter $q_0$ directly controls the exploration-exploitation balance:

| $q_0$ | Behaviour |
|-------|----------|
| 0 | Pure exploration (like AS) |
| 1 | Pure exploitation (greedy) |
| 0.9 | 90% greedy, 10% exploratory (typical ACS value) |

**Note**: In ACS, $\alpha = 1$ (pheromone exponent is always 1, no $\alpha$ parameter).

### Innovation 2: Global Pheromone Update (Elite Strategy)

In ACS, **only the best-so-far ant** updates pheromone after each iteration:

**Step 1: Evaporation (only on best-so-far tour)**
$$\tau_{ij} \leftarrow (1 - \rho) \cdot \tau_{ij} \quad \forall (i,j) \in T_{bs}$$

**Step 2: Deposition (only on best-so-far tour)**
$$\tau_{ij} \leftarrow \tau_{ij} + \rho \cdot \Delta\tau_{ij}^{bs}$$

Where:
$$\Delta\tau_{ij}^{bs} = \begin{cases} 1/L_{bs} & \text{if } (i,j) \in T_{bs} \\ 0 & \text{otherwise} \end{cases}$$

- $T_{bs}$ = the **best-so-far tour** (the best tour found since the start of the algorithm)
- $L_{bs}$ = length of the best-so-far tour

**Why this works**: By concentrating pheromone on the globally best tour, ACS directs search toward the most promising region of the search space.

### Innovation 3: Local Pheromone Update (Online Update)

**During construction** (while ants are building their tour), each ant **reduces** pheromone on edges it just used:

$$\tau_{ij} \leftarrow (1 - \xi) \cdot \tau_{ij} + \xi \cdot \tau_0$$

Where:
- $\xi \in (0, 1)$ is the **local evaporation rate** (typically 0.1)
- $\tau_0$ = initial pheromone value

**Why does local update help?** It makes an edge **less attractive** immediately after it's been used. This encourages other ants in the same iteration to explore different edges, **increasing diversity**.

Think of it as: "I've been here, so let me reduce the sign pointing here so others explore elsewhere."

**Mathematical effect**: With local update, $\tau_{ij}$ can never fall below $\xi \cdot \tau_0 / (\xi) = \tau_0$ times a factor. The minimum pheromone level ensures the graph remains connected.

### Complete ACS Algorithm

```
ALGORITHM Ant_Colony_System(n, m, β, ρ, ξ, q_0, τ_0, max_iter):
    
    // Initialisation
    Compute L_nn (nearest-neighbour tour length)
    τ_0 ← 1 / (n · L_nn)
    τ_ij ← τ_0    for all edges (i,j)
    best_length ← ∞
    
    FOR iter = 1 TO max_iter:
        
        // SOLUTION CONSTRUCTION
        FOR ant k = 1 TO m:
            Place ant k at random starting city
            WHILE tour not complete:
                q ← random(0, 1)
                IF q ≤ q_0:                          // EXPLOITATION
                    j ← argmax_{l ∈ J_i^k} [τ_il · η_il^β]
                ELSE:                                 // EXPLORATION
                    j ← sample from p_ij^k
                
                Move to city j
                // LOCAL PHEROMONE UPDATE
                τ_ij ← (1-ξ)·τ_ij + ξ·τ_0
            
            Complete tour; compute L_k
            Update best-so-far if L_k < best_length
        
        // APPLY LOCAL SEARCH (2-opt or 3-opt)
        FOR ant k = 1 TO m:
            Apply 2-opt to ant k's tour
        Update best-so-far after local search
        
        // GLOBAL PHEROMONE UPDATE (only best-so-far)
        FOR edges (i,j) in T_bs:
            τ_ij ← (1-ρ)·τ_ij + ρ·(1/L_bs)
    
    RETURN best_tour, best_length
```

### ACS Parameters Summary

| Parameter | Symbol | Typical Value | Role |
|-----------|--------|---------------|------|
| Number of ants | $m$ | 10 | Colony size |
| Heuristic exponent | $\beta$ | 2 | Greediness weight |
| Global evaporation | $\rho$ | 0.1 | Learning rate |
| Local evaporation | $\xi$ | 0.1 | Diversity mechanism |
| Exploitation param | $q_0$ | 0.9 | Exploration–exploitation |
| Initial pheromone | $\tau_0$ | $1/(n \cdot L_{nn})$ | Starting pheromone |

### Why ACS Outperforms AS

1. **More directed search**: Only the best solution deposits pheromone → faster convergence
2. **Better diversity**: Local update prevents all ants from following the same path
3. **Faster**: Fewer pheromone updates per iteration
4. **Direct exploitation**: $q_0 = 0.9$ means 90% of decisions are deterministically greedy

**Experimental results** (from Dorigo & Gambardella 1997):
- On Eil51 (51 cities), ACS finds optimal or near-optimal tours consistently
- On kroA100 (100 cities), ACS outperforms AS by over 5% in solution quality
- ACS + local search gives results within **0.3%** of optimal for 100-city instances

## 2.7 MAX-MIN Ant System (MMAS)

**MMAS** (Stützle & Hoos, 1997, 2000) is another major improvement over AS.

### Key Ideas of MMAS

#### 1. Pheromone Bounds

MMAS restricts pheromone values to an interval $[\tau_{min}, \tau_{max}]$:

$$\tau_{min} \leq \tau_{ij} \leq \tau_{max}$$

After every update, pheromone values are clipped:
$$\tau_{ij} \leftarrow \max(\tau_{min}, \min(\tau_{max}, \tau_{ij}))$$

**Why?** This prevents premature convergence (all ants following the same path) by ensuring:
- $\tau_{max}$: No edge can become infinitely dominant
- $\tau_{min}$: No edge is completely abandoned (diversity maintained)

#### 2. Only Best Ant Updates (Elite Strategy)

Like ACS, only the best ant (either iteration-best or global-best) updates pheromone:

$$\tau_{ij} \leftarrow (1-\rho) \cdot \tau_{ij} + \Delta\tau_{ij}^{best}$$

MMAS alternates between using the **iteration-best** and **global-best** ant, which provides more exploration than pure global-best updates.

#### 3. Pheromone Reinitialisation

If stagnation is detected (all ants follow the same path), MMAS **reinitialises** all pheromone trails to $\tau_{max}$, restarting the search.

### MMAS Theoretical Bounds

The maximum pheromone value:

$$\tau_{max} = \frac{1}{\rho \cdot L^*}$$

Where $L^*$ is the estimated optimal tour length (often approximated by the best-so-far).

The minimum pheromone value:

$$\tau_{min} = \frac{\tau_{max}(1 - \sqrt[n]{p_{best}})}{(\text{avg} - 1)\sqrt[n]{p_{best}}}$$

Where $p_{best}$ is the desired probability of constructing the optimal solution, and avg is the average number of choices per step.

## 2.8 Vehicle Routing Problems (VRP)

The **Vehicle Routing Problem** is arguably the most important combinatorial optimization problem in logistics. Prof. Gambardella's ACS work on VRP is landmark research.

### Problem Definition

**Given**:
- A central **depot** (location 0)
- $n$ **customers** with locations and **demands** $q_i$
- A fleet of **vehicles** with capacity $Q$
- Travel costs $d_{ij}$ between all locations

**Find**: A set of routes such that:
1. Every customer is visited exactly once
2. Each route starts and ends at the depot
3. The total demand on each route does not exceed capacity $Q$
4. The **total travel cost is minimised**

### Mathematical Formulation

Let $x_{ij}^v \in \{0, 1\}$ indicate if vehicle $v$ travels from $i$ to $j$.

$$\text{Minimise} \quad \sum_{v} \sum_{i} \sum_{j} d_{ij} x_{ij}^v$$

Subject to:
$$\sum_v \sum_j x_{ij}^v = 1, \quad \forall i \in \{1,\ldots,n\} \quad \text{(each customer visited once)}$$

$$\sum_i q_i \cdot \sum_j x_{ij}^v \leq Q, \quad \forall v \quad \text{(capacity constraint)}$$

$$\sum_j x_{0j}^v = \sum_i x_{i0}^v = 1, \quad \forall v \quad \text{(routes start/end at depot)}$$

### VRP Variants

| Variant | Additional Constraint |
|---------|----------------------|
| CVRP | Capacitated VRP (basic version) |
| VRPTW | Time windows, customers have delivery time ranges |
| VRPPD | Pick-up and delivery |
| MDVRP | Multiple depots |
| SDVRP | Split deliveries allowed |
| Dynamic VRP | Requests arrive in real time |

### Relationship to TSP

VRP generalises TSP:
- If all vehicles have infinite capacity and there's only one vehicle: VRP = TSP
- VRP is NP-hard (since TSP is NP-hard and TSP reduces to VRP)
- VRP search space is **much larger** than TSP of the same size

### Why VRP Matters in Practice

Every logistics company solves VRP daily:
- **UPS** saves 10 million gallons of fuel per year by optimising routes
- **Amazon** routes millions of packages per day
- **Food delivery** (Deliveroo, UberEats) optimises pickup and delivery routes
- **Waste collection**, **school buses**, **emergency services** all involve VRP variants

**Annual economic impact**: Logistics companies spend 50-70% of costs on transportation. A 1% improvement in routing saves billions globally.

## 2.9 ACS for the Capacitated VRP (CVRP)

**Gambardella, Taillard & Agazzi (1999)** extended ACS to VRP, a landmark paper in the field.

### Construction Graph for VRP

For VRP, the construction graph is:
- **Nodes**: Depot (0) + $n$ customer locations
- **Edges**: All pairs, with pheromone $\tau_{ij}$ and heuristic $\eta_{ij} = 1/d_{ij}$

### Modified ACS Transition Rule for VRP

An ant building a VRP solution maintains a **current route** with remaining capacity. When choosing the next customer $j$ to visit from current location $i$:

$$j = \begin{cases} \arg\max_{l \in J_i^k} \left[ \tau_{il} \cdot \eta_{il}^\beta \right] & \text{if } q \leq q_0 \\ \text{Sample from } p_{ij}^k & \text{if } q > q_0 \end{cases}$$

Where $J_i^k$ is now the set of **feasible unvisited customers**, those with demand $q_j$ not exceeding the remaining vehicle capacity:

$$J_i^k = \{j : j \text{ unvisited} \wedge q_j \leq \text{remaining capacity}\}$$

If $J_i^k = \emptyset$ (no feasible customer can be added), the ant **returns to depot**, starts a new route, and continues from the depot.

### VRP-Specific Heuristic Information

A key innovation for VRP is using a **savings-based heuristic** (Clark-Wright savings):

$$\eta_{ij} = \frac{1}{d_{ij}}$$

But more sophisticated heuristics combine distance with **savings**:

$$s_{ij} = d_{0i} + d_{0j} - d_{ij}$$

where $d_{0i}$ is the distance from depot to customer $i$. Savings $s_{ij}$ measures how much we save by combining customers $i$ and $j$ in one route instead of two separate routes.

### Pheromone Update for VRP

The pheromone update in ACS-VRP is the same as in ACS-TSP, but now applies to the **complete set of routes** forming the best VRP solution:

$$\tau_{ij} \leftarrow (1-\rho) \cdot \tau_{ij} + \rho \cdot \frac{1}{L_{bs}}$$

for every edge $(i,j)$ in the best-so-far solution $T_{bs}$ (which is now a set of routes, not a single tour).

The cost $L_{bs}$ is the **total travel distance** of the best-so-far solution:

$$L_{bs} = \sum_{r \in R_{bs}} \text{cost}(r)$$

where $R_{bs}$ is the set of routes in the best solution.

### Two-Phase ACS for VRP (Gambardella et al.)

A particularly effective approach uses **two separate pheromone matrices**:

**Phase 1**: Use ACS to minimise the **number of routes** (vehicles used)
$$\min |R|$$

**Phase 2**: Fix the number of routes; use ACS to minimise **total distance**
$$\min L = \sum_{r \in R} \text{cost}(r)$$

This hierarchical approach mirrors real logistics priorities:
1. First, use as few vehicles as possible (fixed cost)
2. Then, make each vehicle's route as efficient as possible (variable cost)

### Local Search Integration

ACS for VRP achieves its best results when combined with **local search**:

#### 2-opt (intra-route)
Remove two edges within a route and reconnect differently:

$$\text{Route: } \ldots - i - a - b - \ldots - c - d - j - \ldots$$
$$\text{After 2-opt: } \ldots - i - a - c - \ldots - b - d - j - \ldots$$

Accept if the new route is shorter.

#### Or-opt (inter-route)
Move a customer (or a sequence of customers) from one route to another.

#### 3-opt
Remove three edges and reconnect. More powerful than 2-opt but $O(n^3)$ time.

## 2.10 ACO Convergence and Parameter Tuning

### Convergence Analysis

**Theorem (Stützle & Dorigo, 2002)**: If at each iteration there is a nonzero probability of finding the optimal solution, and if the best-so-far solution is used for pheromone update, then ACO converges to the optimal solution with probability approaching 1 as the number of iterations approaches infinity:

$$P(\text{finding optimal by iteration } t) \rightarrow 1 \quad \text{as } t \rightarrow \infty$$

**Important caveat**: This is an asymptotic result. In practice, ACO is run for a fixed number of iterations, and optimality is not guaranteed.

### Pheromone Stagnation

**Stagnation** occurs when all ants follow the same path, pheromone diversity collapses:

$$\tau_{ij} \rightarrow \tau_{max} \text{ for edges in best tour}, \quad \tau_{ij} \rightarrow \tau_{min} \text{ for others}$$

Stagnation indicators:
- **Pheromone entropy** approaches 0
- All ants construct identical (or very similar) tours
- No improvement for many iterations

Stagnation prevention strategies:
1. MMAS pheromone bounds $[\tau_{min}, \tau_{max}]$
2. Pheromone re-initialisation
3. ACS local pheromone update
4. Multiple ant colonies with different parameters

### Parameter Sensitivity Analysis

#### Effect of $\rho$ (evaporation rate)

The **effective search radius** (number of iterations before pheromone becomes negligible):

$$t_{eff} \approx \frac{1}{\rho}$$

For $\rho = 0.1$: $t_{eff} \approx 10$ iterations of "memory"
For $\rho = 0.01$: $t_{eff} \approx 100$ iterations of "memory"

#### Effect of $\beta$ (heuristic exponent)

The relative influence of pheromone vs. heuristic on edge selection:

$$\text{Pheromone dominates when: } \tau_{ij} > \eta_{ij}^\beta / \tau_{il} \cdot \eta_{il}^\beta$$

High $\beta$ → greedy, exploitative, fast convergence but may miss global optimum
Low $\beta$ → pheromone-driven, more exploratory

### The Candidate List Strategy

A critical efficiency improvement: instead of considering ALL $n-1$ unvisited cities at each step, only consider the **$cl$ nearest cities** (candidate list).

$$J_i^k \leftarrow \{j : j \text{ is among the } cl \text{ nearest unvisited cities to } i\}$$

If the candidate list is empty (all $cl$ nearest cities are visited), fall back to considering all remaining cities.

**Effect on complexity**: Reduces per-step computation from $O(n)$ to $O(cl)$ where $cl \ll n$. Typical $cl = 20$ works well even for large instances.

This speeds up ACS by a factor of $n/cl$ with minimal quality loss.

### Q1: What is stigmergy, and why is it a powerful coordination mechanism?

**Answer**:

**Stigmergy** (from Greek *stigma* = mark, *ergon* = work) is **indirect coordination through environmental modification**.

Unlike direct communication (ant A tells ant B where food is), stigmergy works through the environment:
- Agent A modifies the environment (deposits pheromone)
- Agent B senses the environment (detects pheromone) and modifies its behaviour accordingly
- Agent A and B never directly interact

**Why is it powerful**?
1. **Scalable**: Works with thousands of agents simultaneously without communication bottlenecks
2. **Robust**: The death or failure of individual agents doesn't crash the system
3. **Emergent**: Complex collective behaviour from simple local rules
4. **Self-reinforcing**: Good solutions accumulate more signals, making them more likely to be found again

The pheromone trail acts as a **distributed memory**, the collective "knowledge" of the colony is stored in the environment, not in any individual ant.

---

### Q2: Derive the steady-state pheromone level on an edge that is consistently reinforced by the best ant in ACS.

**Answer**:

Consider an edge $(i,j)$ that belongs to the best-so-far tour $T_{bs}$ with cost $L_{bs}$.

The ACS global update rule is:
$$\tau_{ij}^{(t+1)} = (1-\rho) \tau_{ij}^{(t)} + \rho \cdot \frac{1}{L_{bs}}$$

At **steady state**, $\tau_{ij}^{(t+1)} = \tau_{ij}^{(t)} = \tau^*$:

$$\tau^* = (1-\rho)\tau^* + \frac{\rho}{L_{bs}}$$

$$\tau^* - (1-\rho)\tau^* = \frac{\rho}{L_{bs}}$$

$$\rho \tau^* = \frac{\rho}{L_{bs}}$$

$$\boxed{\tau^* = \frac{1}{L_{bs}}}$$

The steady-state pheromone level on the best-solution edges equals the **inverse of the best-so-far tour length**. Shorter tours → higher pheromone levels → stronger attraction.

---

### Q3: What happens in ACS when $q_0 = 1$? What are the implications?

**Answer**:

When $q_0 = 1$, all decisions are exploitation:
$$j = \arg\max_{l \in J_i^k} [\tau_{il} \cdot \eta_{il}^\beta], \quad \text{always}$$

**Implications**:
1. **All ants build identical tours** (given the same starting city and pheromone state)
2. **No exploration**, once a suboptimal tour is found, the algorithm is trapped
3. The pheromone update concentrates all pheromone on one tour
4. The local update still provides some diversity (it reduces attractiveness of used edges)

In practice, $q_0 = 1$ typically gives worse results than $q_0 = 0.9$ because the $10\%$ exploration is crucial for escaping local optima.

**Analogy**: A student who always goes to the library they've rated highest never discovers potentially better libraries.

---

### Q4: Explain why the local pheromone update in ACS promotes diversity, using a concrete example.

**Answer**:

**Without local update**: Suppose 10 ants all start building their tour simultaneously. Early on, edge $(A,B)$ has high pheromone. Many ants will choose edge $(A,B)$, they all converge on similar tours, wasting diversity.

**With local update**: The first ant to traverse $(A,B)$ immediately reduces $\tau_{AB}$:
$$\tau_{AB} \leftarrow (1-\xi)\tau_{AB} + \xi \cdot \tau_0$$

Since $(1-\xi) < 1$, the pheromone drops. The second ant to consider $(A,B)$ sees less pheromone. After $k$ ants use the edge, pheromone is:
$$\tau_{AB}^{(k)} = (1-\xi)^k \tau_{AB}^{(0)} + \tau_0 \left(1 - (1-\xi)^k\right)$$

This approaches $\tau_0$ as $k \rightarrow \infty$, the edge becomes no more attractive than any unexplored edge.

**Result**: Later ants are pushed toward less-used edges, exploring different parts of the search space. This is equivalent to sampling **without replacement** at the population level.

### Q5: Compare ACS and MMAS. When would you choose one over the other?

**Answer**:

| Feature | ACS | MMAS |
|---------|-----|------|
| Transition rule | Pseudo-random proportional (with $q_0$) | Standard proportional |
| Best update | Global-best only | Alternates iteration-best / global-best |
| Diversity mechanism | Local pheromone update | Pheromone bounds + reinitialisation |
| Stagnation recovery | Local update (gradual) | Explicit reinitialisation (aggressive) |
| Parameters | $q_0, \xi, \rho, \beta$ | $\rho, \tau_{min}, \tau_{max}$ |

**Choose ACS when**:
- You want simple, well-understood parameter tuning
- The problem benefits from greedy exploitation (dense local optima structure)
- You plan to combine with aggressive local search
- Problems like TSP and VRP where $q_0$ tuning gives large gains

**Choose MMAS when**:
- The search landscape is rugged and diverse exploration is critical
- Stagnation is a significant risk
- You want provable theoretical properties (bounds guarantee diversity)
- Competition benchmarks (MMAS holds or held many world records on TSP benchmarks)

**In practice**: MMAS + 3-opt local search is among the best known algorithms for TSP. ACS + local search is typically better for VRP.

### Q6: A VRP instance has 50 customers, each with demand 10. The vehicle capacity is 50. What is the minimum number of routes needed? Can ACS always find this minimum?

**Answer**:

**Minimum routes** (lower bound from capacity):

Total demand = $50 \times 10 = 500$
Each vehicle can carry at most 50 units.
$$\text{Min routes} \geq \left\lceil \frac{\text{Total demand}}{\text{Capacity}} \right\rceil = \left\lceil \frac{500}{50} \right\rceil = 10$$

So at least 10 routes are needed. Note this is a lower bound, geographic constraints may require more.

**Can ACS always find this minimum?** Not guaranteed.

ACS is a heuristic, it may produce 11 or 12 routes in practice. The two-phase approach (first minimise routes, then minimise distance) helps, but doesn't guarantee optimality. The two-phase ACS by Gambardella et al. achieves the minimum number of routes on most standard benchmark instances, but occasionally needs one extra vehicle.

## 3.1 Biological Inspiration: Darwin's Evolution

### The Theory of Natural Selection

Darwin's theory (1859) contains three key observations that inspire GAs:

**1. Variation**: Individuals in a population differ in their traits (genotype/phenotype)

**2. Heredity**: Traits are passed from parents to offspring (with some variation through mutation and recombination)

**3. Selection**: Individuals better adapted to their environment survive and reproduce more (survival of the fittest)

**Result**: Over generations, the population becomes better adapted to its environment, **evolution is a search algorithm**!

### Biological vs. Algorithmic Analogy

| Biology | Genetic Algorithm |
|---------|-------------------|
| Organism | Candidate solution |
| Chromosome | Encoding of solution (bitstring, etc.) |
| Gene | Individual element of the encoding |
| Allele | Value of a gene |
| Fitness | Objective function value $f(s)$ |
| Population | Set of candidate solutions |
| Selection | Choosing parents for reproduction |
| Crossover | Combining parent solutions |
| Mutation | Random modification of a solution |
| Generation | One iteration of the GA |

### What Makes GA Different from ACO?

- **ACO**: Constructs solutions from scratch, guided by pheromone memory
- **GA**: Maintains a population of complete solutions, improves through recombination

GA is a **population-based, evolutionary algorithm**, it simulates the entire evolutionary process.

## 3.2 The Basic Genetic Algorithm (Holland, 1975)

### Algorithm Overview

```
ALGORITHM Basic_GA(pop_size, p_c, p_m, max_gen):
    
    // Initialisation
    P(0) ← generate random population of pop_size individuals
    Evaluate fitness f(s) for all s ∈ P(0)
    
    FOR gen = 1 TO max_gen:
        P'(gen) ← empty population
        
        WHILE |P'(gen)| < pop_size:
            // Selection
            p1 ← SelectParent(P(gen))
            p2 ← SelectParent(P(gen))
            
            // Crossover
            IF random() < p_c:
                (c1, c2) ← Crossover(p1, p2)
            ELSE:
                (c1, c2) ← (p1, p2)    // copy parents
            
            // Mutation
            c1 ← Mutate(c1, p_m)
            c2 ← Mutate(c2, p_m)
            
            // Add to new population
            P'(gen) ← P'(gen) ∪ {c1, c2}
        
        P(gen) ← P'(gen)
        Evaluate fitness for all s ∈ P(gen)
    
    RETURN argmax_{s ∈ P(max_gen)} f(s)
```

### Key Parameters

| Parameter | Symbol | Typical Value | Role |
|-----------|--------|---------------|------|
| Population size | $N$ | 50–200 | Diversity |
| Crossover probability | $p_c$ | 0.6–0.9 | Recombination rate |
| Mutation probability | $p_m$ | $1/L$ to $0.1$ | Exploration |
| Generations | $G$ | 100–10000 | Computation budget |

Where $L$ is the chromosome length.

### The GA Cycle

```
[Initial Population]
        ↓
    [Evaluation] ← f(s) for all individuals
        ↓
    [Selection] ← Better individuals reproduce more
        ↓
    [Crossover] ← Combine good solutions
        ↓
    [Mutation]  ← Introduce variation
        ↓
    [New Population]
        ↓
    Repeat until termination
```

## 3.3 Representation and Encoding

**The most important design decision in a GA is how to represent solutions.**

The representation must be:
1. **Complete**: Every valid solution must be representable
2. **Closed under operators**: Crossover and mutation should produce valid solutions
3. **Efficient**: Fitness evaluation should be fast

### Binary Encoding

Each solution is a **bitstring** of length $L$:
$$s = (b_1, b_2, \ldots, b_L), \quad b_i \in \{0, 1\}$$

**Example**: Knapsack problem with $n=5$ items:
- Solution: Which items to include
- Encoding: $s = (1, 0, 1, 1, 0)$ = include items 1, 3, 4

Search space size: $2^L$ (exponential in $L$).

For continuous variables, a real value $x \in [a, b]$ can be encoded as a $k$-bit integer:
$$x = a + \frac{(b-a)}{2^k - 1} \cdot \text{int}(b_1 b_2 \ldots b_k)$$

where $\text{int}(b_1 \ldots b_k) = \sum_{i=1}^k b_i \cdot 2^{k-i}$.

**Precision**: With $k=16$ bits and range $[-5, 5]$, resolution is $\Delta x = 10/(2^{16}-1) \approx 0.00015$.

### Real-Valued Encoding

For continuous optimization, directly encode real values:
$$s = (x_1, x_2, \ldots, x_n), \quad x_i \in [l_i, u_i] \subset \mathbb{R}$$

**Advantages**: Avoids Hamming cliff problem, natural representation, easier arithmetic crossover.

### Permutation Encoding

For problems requiring orderings (TSP, scheduling, VRP):
$$s = (\pi(1), \pi(2), \ldots, \pi(n))$$
where $\pi$ is a permutation of $\{1, 2, \ldots, n\}$.

**Challenge**: Standard crossover breaks the permutation property, need special operators (PMX, OX, etc.).

**Example** (TSP, 6 cities):
$$s = (3, 1, 5, 2, 6, 4) \quad \text{(visit cities in this order)}$$

### The Hamming Cliff Problem

A subtle issue with binary encoding: **adjacent integer values can have very different bit representations**.

Standard binary:
- 7 = 0111
- 8 = 1000

These are adjacent numbers but differ in **all 4 bits**! A mutation that flips one bit of 0111 gives: 0110 (=6), 0101 (=5), 0011 (=3), 1111 (=15), it can jump to 15 but cannot easily reach 8.

**Solution**: Use **Gray encoding** where adjacent integers differ in exactly one bit:
- 7 = 0100 (Gray)
- 8 = 1100 (Gray)

Gray code conversion: $g_i = b_i \oplus b_{i-1}$ (XOR with previous bit).

## 3.4 Selection Mechanisms

Selection determines which individuals **reproduce** (become parents). It creates **selection pressure** toward better solutions.

### 1. Fitness-Proportional Selection (Roulette Wheel)

Individual $i$ is selected with probability proportional to its fitness:

$$P(\text{select } i) = \frac{f(s_i)}{\sum_{j=1}^{N} f(s_j)}$$

**Implementation** (roulette wheel):
1. Compute cumulative fitness: $F_k = \sum_{i=1}^k f(s_i)$
2. Draw $r \sim U[0, F_N]$
3. Select individual $k$ such that $F_{k-1} < r \leq F_k$

**Problem**: Susceptible to **super-individual** domination. If one individual has fitness $100\times$ the average, it will dominate selection, reducing diversity.

Also: If all fitnesses are similar (converged population), selection pressure vanishes.

### 2. Tournament Selection

1. Randomly pick $k$ individuals (tournament size, typically $k=2$)
2. Select the **best individual** from the tournament

With $k=2$ (binary tournament):
$$P(\text{select } i) = \frac{2}{N} \cdot \text{Rank}(i) / N \quad \text{(approximately)}$$

**Advantages**:
- Simple, efficient: $O(k)$ per selection
- Robust to fitness scaling
- Tunable pressure via $k$ (larger $k$ = more pressure)

**Most popular** selection mechanism in practice.

### 3. Rank Selection

Sort individuals by fitness; assign selection probability based on **rank**, not raw fitness:

$$P(\text{select } i) = \frac{2 \cdot \text{rank}(i)}{N(N+1)}$$

Where rank(i) = 1 for worst, $N$ for best.

**Advantage**: Uniform selection pressure regardless of fitness scale. Prevents super-individual domination.

**Disadvantage**: Ignores magnitudes of fitness differences, can be too slow.

### 4. Truncation Selection

Keep only the top $T\%$ of individuals; randomly select parents from these.

Used in evolution strategies (e.g., $(\mu + \lambda)$-ES).

### Elitism

**Elitism**: Always copy the best individual(s) from generation $t$ to generation $t+1$.

**Why?** Without elitism, the best solution can be destroyed by crossover or mutation, elitism ensures monotonic improvement.

**Theorem**: A GA with elitism (keeping at least one best individual) converges to the global optimum with probability 1 under mild conditions (sufficient mutation, infinite time).

## 3.5 Crossover Operators

Crossover (recombination) combines two parent solutions to create offspring. It is the **primary exploitation operator** in GAs.

### One-Point Crossover (Binary)

1. Choose a random crossover point $c \in \{1, \ldots, L-1\}$
2. Offspring 1 = first $c$ bits of parent 1 + last $L-c$ bits of parent 2
3. Offspring 2 = first $c$ bits of parent 2 + last $L-c$ bits of parent 1

**Example** ($L=8$, $c=5$):
```
Parent 1:  10110 | 100
Parent 2:  01001 | 011
           ─────┼────
Child 1:   10110   011
Child 2:   01001   100
```

**Mathematical formulation**: For a cutpoint $c$:
$$\text{Child}_1 = (p_1[1:c], p_2[c+1:L])$$
$$\text{Child}_2 = (p_2[1:c], p_1[c+1:L])$$

### Two-Point Crossover

Choose two cut points $c_1 < c_2$. Exchange the segment between them.

**Example** ($c_1=2$, $c_2=6$):
```
Parent 1:  10 | 1101 | 00
Parent 2:  01 | 0010 | 11
Child 1:   10   0010   00
Child 2:   01   1101   11
```

### Uniform Crossover

For each position $i$, independently:
$$\text{Child}_1[i] = \begin{cases} p_1[i] & \text{with prob. } 0.5 \\ p_2[i] & \text{with prob. } 0.5 \end{cases}$$

**Properties**: More disruptive than one-point. Mixes genes from both parents uniformly. Good for problems where gene interactions are minimal.

### Arithmetic Crossover (Real-Valued)

For real-valued encoding, create offspring as **linear combination** of parents:

$$\text{Child}_1 = \lambda \cdot p_1 + (1-\lambda) \cdot p_2$$
$$\text{Child}_2 = (1-\lambda) \cdot p_1 + \lambda \cdot p_2$$

where $\lambda \sim U[0, 1]$ (or a fixed value like 0.5).

**BLX-$\alpha$ Crossover** (Blend Crossover):
For each dimension $i$, create offspring in range $[x_{\min} - \alpha \cdot I, x_{\max} + \alpha \cdot I]$ where:
- $x_{\min} = \min(p_1[i], p_2[i])$
- $x_{\max} = \max(p_1[i], p_2[i])
- $I = x_{\max} - x_{\min}$
- $\alpha \in [0, 0.5]$ (typical: 0.5)

This allows offspring to explore **beyond** the interval between parents.

### Order Crossover (OX), for Permutations (TSP)

**Problem with standard crossover on permutations**: It creates duplicates!

Example (TSP, 6 cities):
```
Parent 1: [1 2 3 4 5 6]
Parent 2: [4 2 6 1 3 5]
One-point: [1 2 3 | 1 3 5] → City 1 and 3 appear twice! INVALID
```

**Order Crossover (OX)** preserves the relative order:
1. Select a random segment from parent 1
2. Copy segment to offspring
3. Fill remaining positions with cities from parent 2, **in the order they appear**, skipping already-included cities

**Example** (segment positions 3-5):
```
Parent 1: [1 2 | 3 4 5 | 6]
Parent 2: [4 2 6 1 3 5]

Step 1: Copy segment: [_ _ 3 4 5 _]
Step 2: From parent 2 (starting after segment end, wrapping): 
        Remaining cities in parent 2 order: 1 (yes), 3 (skip-in), 5 (skip-in), 4 (skip-in), 2 (yes), 6 (yes)
        → [1, 2, 6] fill remaining positions left to right
Step 3: Child: [1 2 3 4 5 6]
```

### Partially Mapped Crossover (PMX)

PMX creates a mapping between parents to resolve conflicts:

1. Select a segment from parent 1; copy to offspring
2. For each gene $g$ in the parent 2 segment that isn't in the copied segment:
   - Find where $g$'s position in offspring is occupied by gene $g'$
   - Find where $g'$ is in parent 2, place $g$ there

PMX preserves **absolute positions** better; OX preserves **relative order** better.

## 3.6 Mutation Operators

Mutation is the **primary exploration operator**, it introduces new genetic material that cannot be obtained through crossover alone.

### Bit-Flip Mutation (Binary Encoding)

For each bit $b_i$, flip with probability $p_m$:
$$b_i \leftarrow \begin{cases} 1 - b_i & \text{with prob. } p_m \\ b_i & \text{with prob. } 1-p_m \end{cases}$$

**Expected number of flips**: $L \cdot p_m$

**Standard setting**: $p_m = 1/L$ → expected 1 bit flipped per chromosome.

With $p_m = 1/L$, the probability that a chromosome is unchanged:
$$P(\text{no flip}) = (1 - 1/L)^L \approx e^{-1} \approx 0.368$$

So about 63% of chromosomes receive at least one mutation, a healthy level.

### Gaussian Mutation (Real-Valued)

Add Gaussian noise to each dimension:
$$x_i \leftarrow x_i + \mathcal{N}(0, \sigma^2)$$

Where $\sigma$ is the **mutation step size** (key parameter!).

**Self-adaptive mutation**: Let $\sigma$ itself evolve! Each individual carries its own $\sigma$:
$$\sigma' = \sigma \cdot e^{\tau \mathcal{N}(0,1)}$$
$$x_i' = x_i + \sigma' \mathcal{N}(0,1)$$

where $\tau = 1/\sqrt{n}$ (standard Evolution Strategy setting). Good solutions tend to have good $\sigma$ values, so $\sigma$ adapts automatically!

### Permutation Mutations

#### Swap Mutation
Choose two random positions $i, j$ and swap their values:
$$\pi = (1,2,3,4,5) \xrightarrow{\text{swap } i=2, j=4} (1,4,3,2,5)$$

#### Inversion Mutation (2-opt move)
Choose a segment and reverse it:
$$\pi = (1,2,3,4,5) \xrightarrow{\text{invert } [2:4]} (1,4,3,2,5)$$

This is equivalent to one step of 2-opt local search, very effective for TSP!

#### Insertion Mutation
Remove an element and insert it elsewhere:
$$\pi = (1,2,3,4,5) \xrightarrow{\text{move 3 to position 5}} (1,2,4,5,3)$$

### The Role of Mutation Rate

| $p_m$ | Effect |
|-------|--------|
| Too high (0.5+) | Destroys good solutions; random search |
| Too low (0) | No exploration; premature convergence |
| Just right ($1/L$) | Balances exploration and exploitation |

**Why $1/L$?** Each chromosome has $L$ genes. Setting $p_m = 1/L$ means we expect **exactly one gene** to mutate per chromosome. This maintains diversity without destroying too much information.

## 3.7 Schema Theorem and Building Blocks

This is the **theoretical foundation** of why GAs work (Holland, 1975).

### What Is a Schema?

A **schema** $H$ is a template that describes a subset of chromosomes by specifying fixed values at some positions and wildcards (`*`) at others.

**Example** ($L=6$, binary):
- Schema $H = (1, *, 0, *, *, 1)$ matches all bitstrings with $b_1=1$, $b_3=0$, $b_6=1$
- It matches: $100001$, $110011$, $100101$, $111011$, ... ($2^3 = 8$ strings in total)

### Schema Properties

**Order** $o(H)$: Number of **fixed** positions (non-wildcard characters)
$$o(H = (1,*,0,*,*,1)) = 3$$

**Defining Length** $\delta(H)$: Distance between the **first and last** fixed position
$$\delta(H = (1,*,0,*,*,1)) = 6 - 1 = 5$$

**Fitness of a schema** $\hat{f}(H, t)$: Average fitness of all chromosomes matching $H$ at generation $t$.

### The Schema Theorem

**Holland's Schema Theorem**: The expected number of instances of schema $H$ at generation $t+1$ satisfies:

$$E[m(H, t+1)] \geq m(H, t) \cdot \frac{\hat{f}(H, t)}{\bar{f}(t)} \cdot \left(1 - p_c \cdot \frac{\delta(H)}{L-1}\right) \cdot \left(1 - p_m\right)^{o(H)}$$

Where:
- $m(H, t)$ = number of instances of $H$ in population at time $t$
- $\hat{f}(H, t)$ = average fitness of instances of $H$
- $\bar{f}(t)$ = average fitness of entire population
- $p_c$ = crossover probability
- $\delta(H)$ = defining length of $H$
- $L$ = chromosome length
- $p_m$ = mutation probability

### Interpreting the Schema Theorem

The formula has three factors:

**Factor 1**: $\dfrac{\hat{f}(H, t)}{\bar{f}(t)}$, **selection factor**
- If $H$ has above-average fitness ($\hat{f} > \bar{f}$): $m(H)$ grows
- If $H$ has below-average fitness: $m(H)$ shrinks
- **Selection increases the proportion of above-average schemas**

**Factor 2**: $1 - p_c \cdot \dfrac{\delta(H)}{L-1}$, **crossover survival probability**
- Short defining length $\delta(H) \ll L$: rarely disrupted → approaches 1
- Long defining length: frequently disrupted → approaches $1-p_c$
- **Short schemas survive crossover better**

**Factor 3**: $(1-p_m)^{o(H)}$, **mutation survival probability**
- Low order $o(H)$ (few fixed bits): rarely mutated → approaches 1
- High order: frequently mutated → approaches 0
- **Low-order schemas survive mutation better**

### The Building Block Hypothesis

**Definition**: A **building block** is a short, low-order, above-average schema.

The Schema Theorem says building blocks grow exponentially:

$$m(H, t) \approx m(H, 0) \cdot \left(\frac{\hat{f}(H)}{\bar{f}}\right)^t$$

**The Building Block Hypothesis** (Holland): GAs work by identifying, recombining, and propagating building blocks. Crossover combines good building blocks from different parents into offspring that are better than either parent.

**Example**: In TSP, if "city 5 → city 8" is a good sub-path (short edge), it's a building block. The GA will discover this and increasingly include it in solutions.

### Implicit Parallelism

A population of $N$ chromosomes of length $L$ **implicitly evaluates** approximately $N \cdot 2^{L/3}$ schemas (Mitchell, 1996). This is far more than the $N$ fitness evaluations actually performed!

This **implicit parallelism** is the key computational advantage of GAs over exhaustive search.

## 3.8 Population-Based Incremental Learning (PBIL)

**PBIL** (Baluja, 1994) is a hybrid that blends GAs with Estimation of Distribution Algorithms (EDAs). It is simpler than standard GAs but often equally effective.

### Core Idea

Instead of maintaining an explicit population of chromosomes, PBIL maintains a **probability vector** $\mathbf{p} = (p_1, p_2, \ldots, p_L)$ where:

$$p_i = P(\text{bit } i = 1 \text{ in a good solution})$$

Initially: $p_i = 0.5$ for all $i$ (all bits equally likely), maximum uncertainty.

Over iterations, $\mathbf{p}$ is **learned from good solutions** and updated to reflect the best solutions found.

### PBIL Algorithm

```
ALGORITHM PBIL(L, N, lr, p_mut, μ_shift, max_gen):
    
    // Initialisation
    p_i ← 0.5    for i = 1 to L    // Probability vector
    
    FOR gen = 1 TO max_gen:
        
        // STEP 1: Generate population from probability vector
        FOR k = 1 TO N:
            FOR i = 1 TO L:
                s_k[i] ← 1 with prob. p_i, else 0
            Evaluate f(s_k)
        
        // STEP 2: Find best individual
        s_best ← argmax_{k} f(s_k)
        
        // STEP 3: Update probability vector toward best
        FOR i = 1 TO L:
            p_i ← p_i · (1 - lr) + s_best[i] · lr
        
        // STEP 4: Mutation on probability vector
        FOR i = 1 TO L:
            IF random() < p_mut:
                p_i ← p_i · (1 - μ_shift) + random(0,1) · μ_shift
    
    // Sample final solution
    RETURN best solution found
```

### The PBIL Update Formula

The key update rule moves each probability toward the best solution:

$$p_i \leftarrow p_i (1 - lr) + s_{best}[i] \cdot lr$$

where $lr \in (0, 1)$ is the **learning rate**.

**Expanding this**:
- If $s_{best}[i] = 1$: $p_i \leftarrow p_i + lr(1 - p_i)$ → probability **increases** toward 1
- If $s_{best}[i] = 0$: $p_i \leftarrow p_i - lr \cdot p_i = p_i(1-lr)$ → probability **decreases** toward 0

After $t$ updates all pushing the same direction (say, toward 1):

$$p_i(t) = 1 - (1 - lr)^t(1 - p_i(0))$$

This converges to 1 as $t \rightarrow \infty$ (assuming the best solution always has $s_{best}[i]=1$).

### PBIL Mutation

Mutation perturbs the probability vector to prevent premature convergence:

$$p_i \leftarrow p_i (1 - \mu_{shift}) + \text{Bernoulli}(0.5) \cdot \mu_{shift}$$

This is analogous to GA mutation but operates on the probability model, not individual solutions.

### PBIL vs. Standard GA

| Feature | Standard GA | PBIL |
|---------|------------|------|
| Population storage | $N$ chromosomes of length $L$ | One probability vector of length $L$ |
| Memory | $O(NL)$ bits | $O(L)$ real values |
| Learning | Implicit (schema sampling) | Explicit (probability update) |
| Crossover | Required | Not used |
| Mutation | On chromosomes | On probability vector |
| Interpretability | Low | Higher, $p_i$ shows which bits are "important" |

### Connection to Estimation of Distribution Algorithms

PBIL is one of the first **Estimation of Distribution Algorithms (EDAs)**, algorithms that build explicit probabilistic models of the search space:

$$p(\mathbf{x}) = \prod_{i=1}^{L} p_i^{x_i} (1-p_i)^{1-x_i}$$

This assumes **independence** between variables. More sophisticated EDAs (MIMIC, BOA) model variable **interactions**, giving better performance on problems with dependencies.

## 3.9 Evolution Strategies and CMA-ES

### Evolution Strategies (ES)

ES (Rechenberg, 1965; Schwefel, 1965) was developed independently of GAs, focusing on **continuous optimization**.

### The $(1+1)$-ES

The simplest ES: one parent, one offspring:

$$\mathbf{x}' = \mathbf{x} + \sigma \cdot \mathcal{N}(\mathbf{0}, \mathbf{I})$$

If $f(\mathbf{x}') \leq f(\mathbf{x})$ (minimization): accept $\mathbf{x}'$
Otherwise: keep $\mathbf{x}$

### The $(\mu + \lambda)$-ES and $(\mu, \lambda)$-ES

- $(\mu + \lambda)$: $\mu$ parents generate $\lambda$ offspring; select $\mu$ best from **both** parents and offspring
- $(\mu, \lambda)$: $\mu$ parents generate $\lambda$ offspring; select $\mu$ best from **offspring only** (more exploration)

### The $1/5$ Success Rule (Rechenberg's Rule)

Rechenberg (1973) proved: for a sphere function, the optimal mutation step size gives a **1/5 success ratio**:

$$P(\text{improvement}) \approx \frac{1}{5} \quad \text{(optimal)}$$

Adaptation rule:
- If success rate over last $10n$ steps > $1/5$: increase $\sigma$ by factor $c > 1$
- If success rate < $1/5$: decrease $\sigma$ by factor $c < 1$
- Typical: $c = 1.22$ (Rechenberg) or $c = e^{1/(5n)}$

### Covariance Matrix Adaptation ES (CMA-ES)

CMA-ES (Hansen & Ostermeier, 2001) is the state-of-the-art ES for continuous problems.

**Key idea**: Adapt the **entire covariance matrix** of the mutation distribution, not just the step size.

The mutation distribution at generation $t$:

$$\mathbf{x}' \sim \mathcal{N}(\mathbf{m}^{(t)}, (\sigma^{(t)})^2 \mathbf{C}^{(t)})$$

Where:
- $\mathbf{m}^{(t)}$ = mean of population (current search centre)
- $\sigma^{(t)}$ = overall step size
- $\mathbf{C}^{(t)}$ = covariance matrix (shape of search distribution)

**Mean update** (using best $\mu$ of $\lambda$ offspring):
$$\mathbf{m}^{(t+1)} = \sum_{i=1}^{\mu} w_i \mathbf{x}_{i:\lambda}$$

where $\mathbf{x}_{i:\lambda}$ is the $i$-th best offspring and $w_i > 0$ are weights ($\sum w_i = 1$).

**Covariance matrix update** (rank-one update):
$$\mathbf{C}^{(t+1)} = (1-c_1)\mathbf{C}^{(t)} + c_1 \mathbf{p}_c \mathbf{p}_c^T$$

where $\mathbf{p}_c$ is the **evolution path** (cumulated steps).

**CMA-ES is essentially learning a good search distribution**, the covariance matrix captures the problem's geometry.

## 3.10 Multi-Objective Genetic Algorithms

Many real problems have **multiple conflicting objectives**:
- Minimise cost AND maximise quality
- Minimise travel time AND minimise fuel consumption
- Maximise profit AND minimise risk

### Pareto Optimality

**Definition**: Solution $\mathbf{a}$ **Pareto-dominates** $\mathbf{b}$ (denoted $\mathbf{a} \succ \mathbf{b}$) if:

$$f_i(\mathbf{a}) \leq f_i(\mathbf{b}) \; \forall i \quad \text{AND} \quad f_j(\mathbf{a}) < f_j(\mathbf{b}) \; \text{for some } j$$

(at least as good in all objectives, strictly better in at least one)

**Pareto-optimal solution**: Not dominated by any other feasible solution.

**Pareto front**: The set of all Pareto-optimal solutions.

**Goal of multi-objective optimization**: Find a diverse set of solutions approximating the Pareto front.

### NSGA-II: Non-Dominated Sorting GA II (Deb et al., 2002)

The most widely used multi-objective GA.

**Key concepts**:

**1. Non-dominated sorting**: Sort population into fronts $F_1, F_2, F_3, \ldots$
- $F_1$ = Pareto-optimal solutions (non-dominated by anyone)
- $F_2$ = solutions dominated only by $F_1$
- $F_k$ = solutions dominated only by solutions in $F_1, \ldots, F_{k-1}$

**2. Crowding distance**: Measures how isolated a solution is in objective space:

$$d_i = \sum_{m=1}^{M} \frac{f_m(x_{i+1}) - f_m(x_{i-1})}{f_m^{max} - f_m^{min}}$$

Higher crowding distance → solution is in a less crowded region → preferred to maintain diversity.

**3. Crowded comparison operator**: Solution $i$ is preferred over $j$ if:
- $i$ has better rank (lower front number), OR
- Same rank AND higher crowding distance

### Q1: Why is the mutation probability $1/L$ considered "just right"? Prove it leads to expected 1 mutation.

**Answer**:

**Proof that $p_m = 1/L$ gives expected 1 mutation**:

A chromosome has $L$ bits. Each bit independently mutates with probability $p_m = 1/L$.

The expected number of mutations follows from linearity of expectation:
$$E[\text{# mutations}] = \sum_{i=1}^{L} P(\text{bit } i \text{ flips}) = L \cdot \frac{1}{L} = 1$$

**Why is 1 mutation "just right"?**
- Too many mutations (e.g., $p_m = 0.5$, expected $L/2$ flips) destroy the chromosome's information, it effectively becomes random
- Too few mutations (e.g., $p_m = 0$) provide no exploration, the GA gets stuck
- One mutation per generation balances preserving the parent's information with introducing new variation

The intuition: you want to change one thing at a time. If a solution is nearly optimal, flipping one bit might improve it. Flipping 50 bits nearly guarantees destroying it.

### Q2: Explain the Schema Theorem intuitively. Why are "short" and "low-order" schemas favoured?

**Answer**:

Think of a schema as a **pattern** or **partial solution**. The Schema Theorem says patterns that are:
1. **Above-average** in fitness → grow in the population (natural selection)
2. **Short** (small $\delta(H)$) → rarely destroyed by crossover
3. **Low-order** (small $o(H)$) → rarely destroyed by mutation

**Why short schemas survive crossover**: A crossover point randomly cuts the chromosome. If the fixed positions in a schema span positions 1 and 99 in a length-100 chromosome, almost any crossover point (98 out of 99) would separate them. But if they're at positions 1 and 2, only 1 out of 99 crossover points separates them.

**Why low-order schemas survive mutation**: If a schema has only 2 fixed bits, mutation must hit exactly those 2 bits to destroy it. If it has 20 fixed bits, mutation is much more likely to hit at least one of them.

The **Building Block Hypothesis** says GAs implicitly identify these short, low-order, above-average patterns and combine them, like fitting together Lego blocks to build a complex structure.

### Q3: What is the key difference between PBIL and a standard GA, and what does the probability vector represent?

**Answer**:

**Standard GA**: Maintains an explicit population of $N$ chromosomes (e.g., 100 binary strings of length $L$). Crossover and mutation operate on these chromosomes directly.

**PBIL**: Replaces the entire population with a single **probability vector** $\mathbf{p} = (p_1, \ldots, p_L)$. New solutions are sampled from this distribution; the distribution itself is updated based on the best solutions found.

**What does $p_i$ represent?**
$p_i$ is the probability that bit $i$ equals 1 in a good solution. It captures the **collective knowledge** of the search history:
- $p_i \approx 1.0$: "bit $i$ should almost certainly be 1 in good solutions"
- $p_i \approx 0.0$: "bit $i$ should almost certainly be 0 in good solutions"
- $p_i \approx 0.5$: "bit $i$'s value doesn't matter much" (still uncertain)

**Key difference**: PBIL is an **explicit model learner**, it builds a probabilistic model of what good solutions look like. GA is an **implicit searcher**, it doesn't explicitly represent what it has learned.

### Q4: Why is standard one-point crossover invalid for permutation problems like TSP? Describe OX crossover step by step.

**Answer**:

**Why one-point crossover fails**: TSP requires every city to appear exactly once. Standard crossover takes the first part of parent 1 and the second part of parent 2. But city $c$ might appear in both parts, appearing twice in the offspring, while another city appears zero times.

**Example**:
```
Parent 1: [1 2 3 | 4 5 6 7]  (cut at position 3)
Parent 2: [5 3 7 | 2 4 1 6]
Child:    [1 2 3 | 2 4 1 6]  → City 1 and 2 appear twice! Cities 5,7 missing!
```

**Order Crossover (OX) step by step** (7 cities, cutting positions 3 to 5):

```
Parent 1: [1 2 3 4 5 6 7]
Parent 2: [5 3 7 2 4 1 6]
                ─────
Segment from P1 (positions 3-5): [3 4 5]

Step 1: Copy segment to child:  [_ _ 3 4 5 _ _]

Step 2: Read P2 starting from position 6 (wrapping around):
  P2 sequence from pos 6: [1 6 5 3 7 2 4]
  Remove cities already in segment {3,4,5}: [1 6 7 2]

Step 3: Fill remaining positions left to right:
  Positions to fill: 1, 2, 6, 7
  Values: 1, 6, 7, 2
  
Child 1: [1 6 3 4 5 7 2]
```

All 7 cities appear exactly once. ✓

### Q5: Compare tournament selection vs roulette wheel selection. Under what conditions does each fail?

**Answer**:

**Roulette wheel selection fails when**:
1. **Super-individuals dominate**: If one solution has $f = 1000$ and all others have $f \approx 10$, the super-individual wins 99% of selections. The population loses diversity and gets stuck.
2. **All fitnesses are equal**: Every individual has the same probability → random selection, no pressure toward better solutions. This is especially bad in later generations when the population has converged.

**Tournament selection fails when**:
1. **Tournament size is too large**: If $k=N$ (entire population), we always select the global best → genetic drift, no diversity, stagnation. Equivalent to cloning the best individual.
2. **Tournament size is 1**: Random selection, zero selection pressure.

**Summary**:

| Issue | Roulette Wheel | Tournament |
|-------|---------------|------------|
| Super-individual | Fails badly | Handles well (relative ranking) |
| Uniform fitness | Fails (no pressure) | Works (rank-based) |
| Tuning | None | Via $k$ |
| Computational cost | $O(N)$ to build cumulative | $O(k)$ per selection |

**Practical advice**: Use tournament selection with $k=2$ (binary tournament) as a robust default. It's simple, efficient, and resistant to scale issues.

## 4.1 Local Search: The Foundation

**Local search** is the cornerstone of practical combinatorial optimization. Before understanding Simulated Annealing, Tabu Search, and ILS, you must deeply understand basic local search.

### The Local Search Paradigm

```
PROCEDURE Local_Search(initial_solution s):
    s_current ← s
    REPEAT:
        s_next ← best solution in N(s_current)
        IF f(s_next) < f(s_current):     // improvement
            s_current ← s_next
        ELSE:
            RETURN s_current             // local optimum found
    UNTIL no improvement found
```

### Hill Climbing vs. Steepest Descent

**First Improvement** (Hill climbing): Accept the first neighbour that improves the current solution.
- Faster per iteration (don't need to evaluate full neighbourhood)
- May find lower-quality local optima

**Best Improvement** (Steepest descent): Accept the best neighbour found across the entire neighbourhood.
- Slower per iteration: $O(|\mathcal{N}|)$ evaluations
- Typically finds better local optima

**Random Restart**: To avoid local optima, restart from multiple random initial solutions and return the best local optimum found. Still incomplete (might miss global optimum), but simple and effective.

### The Fundamental Problem

Local search terminates when no improvement is possible in the neighbourhood:

$$f(s^*) \leq f(s), \quad \forall s \in \mathcal{N}(s^*)$$

This is a **local optimum**, not necessarily the global optimum.

**The challenge**: How do we escape local optima?

Three classical strategies:
1. **Accept worse solutions** with some probability (Simulated Annealing)
2. **Forbid returning** to recently visited solutions (Tabu Search)
3. **Perturb** the current solution and restart (Iterated Local Search)

## 4.2 Neighbourhood Structures

The neighbourhood $\mathcal{N}(s)$ is the set of solutions reachable from $s$ in one step. **Defining the right neighbourhood is crucial.**

### For Binary Problems (Knapsack, etc.)

**$k$-bit flip neighbourhood**: All solutions obtained by flipping exactly $k$ bits:
$$\mathcal{N}_k(s) = \{s' : d_H(s, s') = k\}$$

where $d_H$ is the Hamming distance.

- $|\mathcal{N}_1(s)| = L$ (flip one bit)
- $|\mathcal{N}_2(s)| = \binom{L}{2}$ (flip two bits)

### For TSP and VRP: k-opt Neighbourhoods

#### 2-opt Neighbourhood

Remove 2 edges and reconnect the two resulting paths in the only other possible way:

**Before**: $\ldots - i - i+1 - \ldots - j - j+1 - \ldots$
**After**: $\ldots - i - j - \ldots - (i+1) - (j+1) - \ldots$ (reversed middle segment)

**Size**: $|\mathcal{N}_{2opt}(s)| = \binom{n}{2} - n = O(n^2)$

**Improvement delta** (cost of 2-opt move):
$$\Delta = d_{i, j} + d_{i+1, j+1} - d_{i, i+1} - d_{j, j+1}$$

Accept if $\Delta < 0$ (minimization).

**2-opt is extremely effective for TSP**: Reduces tour length by 5–20% over random tours.

#### 3-opt Neighbourhood

Remove 3 edges and reconnect. There are 8 ways to reconnect (some are just combinations of 2-opt moves; the truly new ones are 3-opt moves).

$$|\mathcal{N}_{3opt}(s)| = O(n^3)$$

More powerful but much slower: $O(n^3)$ to evaluate all 3-opt moves.

#### Or-opt Neighbourhood

Move a sequence of $k$ consecutive cities to a different position in the tour:
- **1-move**: Move a single city (also called relocation)
- **2-move**: Move two consecutive cities
- **3-move**: Move three consecutive cities

Or-opt is faster than 3-opt but nearly as effective in practice.

### The $k$-opt Family for TSP Performance

| Method | Avg gap from optimal | Time |
|--------|---------------------|------|
| Nearest neighbour | 20–25% | $O(n^2)$ |
| 2-opt | 5–8% | $O(n^2)$ per pass |
| 3-opt | 3–5% | $O(n^3)$ per pass |
| Lin-Kernighan (LK) | 1–3% | $O(n^{2.2})$ approx |
| LKH (LK-Helsgott) | < 0.5% | $O(n^{2.5})$ approx |

## 4.3 Simulated Annealing (SA)

**SA** (Kirkpatrick, Gelatt & Vecchi, 1983; Černý, 1985) is inspired by the **physical process of annealing in metallurgy**.

### Physical Inspiration

**Annealing** is a heat treatment process:
1. **Heat** a metal to high temperature (atoms move freely, high energy state)
2. **Cool slowly** (atoms gradually settle into low-energy crystalline state)

If you cool **too quickly** (quenching): atoms freeze in a disordered, high-energy state (poor solution quality).
If you cool **slowly enough**: atoms settle into the minimum energy configuration (global optimum).

**In the algorithm**:
- Temperature $T$ = "willingness to accept worse solutions"
- High $T$ → frequent acceptance of worse solutions (exploration)
- Low $T$ → rarely accept worse solutions (exploitation)
- Gradually decrease $T$ → transition from exploration to exploitation

### The Metropolis Criterion

When a neighbour $s'$ with $f(s') > f(s)$ (worse, for minimization) is found:

$$P(\text{accept } s') = e^{-\Delta f / T}$$

where $\Delta f = f(s') - f(s) > 0$ is the deterioration.

**Properties**:
- If $\Delta f = 0$: Accept with probability 1 (always accept equal solutions)
- If $\Delta f > 0$: Accept with probability $e^{-\Delta f / T}$:
  - High $T$: $e^{-\Delta f / T} \rightarrow 1$ (almost always accept)
  - Low $T$: $e^{-\Delta f / T} \rightarrow 0$ (almost never accept)
- If $\Delta f < 0$ (improvement): Accept with probability 1 (always accept improvements)

**Where does $e^{-\Delta f / T}$ come from?** It's the Boltzmann factor from statistical mechanics:

$$P(E) \propto e^{-E / (k_B T)}$$

where $k_B$ is Boltzmann's constant. In SA, $k_B = 1$ (absorbed into $T$).

### The SA Algorithm

```
ALGORITHM Simulated_Annealing(s_0, T_0, T_final, α, inner_iter):
    
    s_current ← s_0     // starting solution
    s_best ← s_0
    T ← T_0             // initial temperature
    
    WHILE T > T_final:
        FOR i = 1 TO inner_iter:    // Markov chain length
            s' ← random neighbour in N(s_current)
            Δf ← f(s') - f(s_current)
            
            IF Δf < 0:             // improvement: always accept
                s_current ← s'
                IF f(s_current) < f(s_best):
                    s_best ← s_current
            ELSE:                   // deterioration: accept with probability
                r ← random(0, 1)
                IF r < exp(-Δf / T):
                    s_current ← s'
        
        T ← α · T          // cooling step
    
    RETURN s_best
```

### Cooling Schedules

The **cooling schedule** determines how $T$ decreases over time.

#### 1. Geometric Cooling (Most Common)

$$T_{k+1} = \alpha \cdot T_k, \quad \alpha \in (0, 1)$$

After $k$ steps: $T_k = \alpha^k \cdot T_0$

Total iterations to reach $T_{final}$:
$$k = \frac{\ln(T_{final}/T_0)}{\ln \alpha} = \frac{\ln(T_0/T_{final})}{-\ln \alpha}$$

Typical: $\alpha = 0.95$ to $0.999$. Slower cooling (higher $\alpha$) → better solutions but longer runtime.

#### 2. Logarithmic Cooling (Theoretically Optimal)

$$T_k = \frac{c}{\ln(1 + k)}$$

**Theorem** (Geman & Geman, 1984): With logarithmic cooling and $c \geq \Delta_{max}$ (maximum energy barrier), SA converges to the global optimum with probability 1.

**Problem**: Convergence is extremely slow, impractical for real problems.

#### 3. Adaptive Cooling

Adjust $T$ based on the acceptance ratio:
$$T_{k+1} = \begin{cases} T_k / (1 + T_k \cdot \ln(1+\delta) / (3\sigma_f)) & \text{Lundy-Mees formula} \end{cases}$$

where $\sigma_f$ is the standard deviation of fitness values.

### Setting the Initial Temperature

The initial temperature $T_0$ should be set so that:
$$P(\text{accept typical worsening move}) \approx \chi_0$$

where $\chi_0 \approx 0.8$ (accept 80% of worsening moves initially).

This gives:
$$T_0 = -\frac{\overline{\Delta f^+}}{\ln \chi_0}$$

where $\overline{\Delta f^+}$ is the average deterioration of worsening moves (estimated by random sampling).

**Example**: If average worsening $\overline{\Delta f^+} = 100$ and we want $\chi_0 = 0.8$:
$$T_0 = -\frac{100}{\ln 0.8} = -\frac{100}{-0.2231} \approx 448$$

### SA Convergence Theorem

**Theorem**: SA simulates a **Markov chain** over the solution space. Under the following conditions:
1. The neighbourhood graph is connected (every state reachable from every other)
2. The cooling schedule satisfies $T_k \geq c / \ln(k+1)$ for some $c$
3. The Metropolis criterion is used for acceptance

SA converges to the **global optimum** with probability 1 as $k \rightarrow \infty$.

**Stationary distribution** of the Markov chain at temperature $T$:

$$\pi_T(s) = \frac{e^{-f(s)/T}}{\sum_{s' \in S} e^{-f(s')/T}} = \frac{1}{Z(T)} e^{-f(s)/T}$$

This is the **Boltzmann distribution**. As $T \rightarrow 0$:

$$\pi_0(s) = \begin{cases} 1/|S^*| & \text{if } s \in S^* \text{ (optimal solutions)} \\ 0 & \text{otherwise} \end{cases}$$

At temperature 0, all probability mass concentrates on the global optima!

## 4.4 Iterated Local Search (ILS)

**ILS** (Lourenço, Martin & Stützle, 2002) is one of the simplest yet most effective meta-heuristics.

### The ILS Idea

Local search terminates at a local optimum. What if we could **systematically escape** local optima?

**ILS Strategy**:
1. Run local search to find a local optimum $s^*_{local}$
2. **Perturb** $s^*_{local}$ to get $s'$ (escape the basin of attraction)
3. Run local search from $s'$ to get a new local optimum $s^{**}_{local}$
4. **Accept or reject** $s^{**}_{local}$ as the new base solution
5. Repeat

### The ILS Algorithm

```
ALGORITHM Iterated_Local_Search(max_iter):
    
    s_0 ← GenerateInitialSolution()
    s* ← LocalSearch(s_0)              // Apply local search
    
    FOR i = 1 TO max_iter:
        s' ← Perturbation(s*, history)  // Escape local optimum
        s'' ← LocalSearch(s')           // Re-optimise
        s* ← AcceptanceCriterion(s*, s'', history)  // Accept or reject
    
    RETURN best solution found
```

### The Four Key Components of ILS

**1. Initial Solution Generation**
- Random initialisation
- Greedy construction heuristic (often better starting point)
- Affects convergence speed but not asymptotic quality

**2. Local Search (the inner loop)**
- Any local search that terminates at a local optimum
- For TSP: 2-opt, 3-opt, Lin-Kernighan
- For CVRP: Or-opt, 2-opt within routes
- **Quality of local search is critical**, better local search → fewer ILS iterations needed

**3. Perturbation** (the key innovation of ILS)
- Must be **strong enough** to escape the basin of attraction
- Must be **weak enough** to preserve the good structure found so far

Too weak → same local optimum found again
Too strong → equivalent to random restart (loses information)

**Perturbation for TSP** (double-bridge move):
A 4-opt move that cannot be undone by 2-opt:
1. Divide tour into 4 segments: $A - B - C - D$
2. Reconnect as: $A - C - B - D$ (or another order)

This creates a tour that is 4-opt different from the original, effectively jumping to a new basin of attraction.

**4. Acceptance Criterion**

Different criteria give different search behaviour:

| Criterion | Rule | Behaviour |
|-----------|------|----------|
| **Greedy** | Accept if $f(s'') < f(s^*)$ | Pure exploitation |
| **Random Walk** | Always accept | Pure exploration |
| **SA-like** | Accept with $e^{-\Delta f/T}$ | Balanced |
| **Restart** | Always restart from $s^*_{best}$ | Focus on global best |

### Why ILS Works

**Theorem** (informal): If the perturbation is ergodic (can reach any solution in finite steps) and the local search always improves solutions, then ILS explores the space of **local optima** rather than the full search space.

For TSP with 100 cities:
- Full search space: $\approx 10^{155}$ tours
- Number of 2-opt local optima: $\approx 10^{10}$ (much smaller!)

ILS performs a random walk over **local optima**, a much smaller, more tractable space.

### ILS for TSP: The Double-Bridge Perturbation

**Why double bridge?** A 2-opt local optimum cannot be improved by any 2-opt move. But the double-bridge move requires removing 4 edges, it creates a tour in a **different 2-opt basin of attraction**.

**Mathematical guarantee**: Any tour reachable from $s$ via the double-bridge that belongs to a different 2-opt basin will not be found by 2-opt alone. ILS with double-bridge systematically explores different basins.

## 4.5 Tabu Search (TS)

**Tabu Search** (Glover, 1986, 1989) uses **short-term memory** to escape local optima without randomness.

### The Core Idea

**Problem**: In local search, once you reach a local optimum, all neighbours are worse. If you accept a worse neighbour to escape, you might immediately go back (cycling).

**TS Solution**: Maintain a **tabu list** of recently visited solutions (or moves). **Forbid** moves that reverse recent steps. This forces the search to explore new territory.

### The Tabu List

Storing full solutions in the tabu list is expensive ($O(N \cdot |s|)$ storage). Instead, store **tabu moves** (attributes of recent moves).

**Example (TSP)**: If we just swapped cities $i$ and $j$, add $(i, j)$ to the tabu list. The move $(j, i)$ (reversing the swap) is now **forbidden** for the next $T_{tabu}$ iterations.

### The TS Algorithm

```
ALGORITHM Tabu_Search(s_0, T_tabu, max_iter):
    
    s_current ← s_0
    s_best ← s_0
    tabu_list ← empty list (length T_tabu)
    
    FOR iter = 1 TO max_iter:
        
        // Find best non-tabu neighbour
        s_next ← argmin_{s' ∈ N(s)} {f(s') : move(s_current → s') ∉ tabu_list
                                               OR f(s') < f(s_best)}  // aspiration
        
        // Update tabu list (FIFO queue)
        Add move(s_current → s_next) to tabu_list
        If |tabu_list| > T_tabu: Remove oldest entry
        
        // Move to best non-tabu neighbour
        s_current ← s_next
        
        // Update best solution
        IF f(s_current) < f(s_best):
            s_best ← s_current
    
    RETURN s_best
```

### The Aspiration Criterion

The aspiration criterion **overrides the tabu list** when a tabu move leads to a solution better than the best found so far:

$$\text{Override tabu if } f(s') < f(s_{best})$$

**Why?** If a tabu move leads to the best solution ever seen, the reason for the tabu restriction (preventing cycling) is no longer relevant, we haven't been to this better solution before.

### The Tabu Tenure $T_{tabu}$

The **tabu tenure** is how long a move stays on the tabu list.

| $T_{tabu}$ | Effect |
|------------|--------|
| Too short (1-2) | Cycling is not prevented |
| Too long (very large) | Too many moves forbidden, search becomes restricted |
| Adaptive | Best performance |

**Typical settings**: $T_{tabu} \in [5, 20]$ for most problems; often set to $\sqrt{n}$ or some fraction of $n$.

**Adaptive tabu tenure**: Randomly select $T_{tabu}$ from $[T_{min}, T_{max}]$ at each step. This prevents periodic cycling patterns.

### Intensification and Diversification

Advanced TS uses longer-term memory for two additional strategies:

**Intensification**: 
- Record the **frequency** of moves leading to good solutions
- Periodically restart from the best solution and favour similar moves
- Implemented via a **frequency memory**: $freq[move]$ = how often this move has been used to reach good solutions

**Diversification**:
- Track **rarely used** moves or **rarely visited** regions
- Periodically force the search into less-explored regions
- Implemented via a **frequency memory**: $freq[move]$ = how often this move has been used
- Add a penalty: $f_{aug}(s') = f(s') + \lambda \cdot \text{freq}[\text{move to } s']$

### Strategic Oscillation

An advanced TS technique: **intentionally cross constraint boundaries**.

In VRP, allow routes to **temporarily exceed capacity** (infeasible solutions), penalised in the objective:

$$f_{aug}(s) = \text{distance}(s) + \lambda \cdot \text{constraint violation}(s)$$

This allows TS to explore regions near feasibility boundaries that often contain the best feasible solutions.

## 4.6 Variable Neighbourhood Search (VNS)

**VNS** (Mladenović & Hansen, 1997) systematically changes the neighbourhood structure to escape local optima.

### The Key Insight

> *"A local minimum in one neighbourhood structure is often not a local minimum in another."*

If we have multiple neighbourhood structures $\mathcal{N}_1, \mathcal{N}_2, \ldots, \mathcal{N}_{k_{max}}$, a solution might be:
- Locally optimal in $\mathcal{N}_1$
- But improvable in $\mathcal{N}_2$!

### VNS Framework

```
ALGORITHM Variable_Neighbourhood_Search:
    s* ← LocalSearch(random solution)
    
    WHILE stopping criterion not met:
        k ← 1
        WHILE k ≤ k_max:
            // SHAKING: random perturbation in k-th neighbourhood
            s' ← RandomSolution(N_k(s*))
            
            // LOCAL SEARCH: find local optimum from s'
            s'' ← LocalSearch(s')
            
            // MOVE OR NOT:
            IF f(s'') < f(s*):
                s* ← s''
                k ← 1        // restart from smallest neighbourhood
            ELSE:
                k ← k + 1    // try next (larger) neighbourhood
    
    RETURN s*
```

### Why VNS Works

The sequence $\mathcal{N}_1 \subset \mathcal{N}_2 \subset \ldots \subset \mathcal{N}_{k_{max}}$ forms a **hierarchy** of neighbourhoods, from small to large.

- Small neighbourhoods: Fast local search, likely local optima
- Large neighbourhoods: Slow but more powerful, escape deeper local optima

VNS starts with cheap escapes and escalates to expensive ones only if needed.

### VNS for TSP

| Neighbourhood | Perturbation |
|--------------|-------------|
| $\mathcal{N}_1$ | 1 random double-bridge move |
| $\mathcal{N}_2$ | 2 random double-bridge moves |
| $\mathcal{N}_3$ | 3 random double-bridge moves |

This is equivalent to ILS with increasing perturbation strength.

## 4.7 Comparing Meta-Heuristics

### Summary Table

| Algorithm | Type | Memory | Randomness | Best For |
|-----------|------|--------|------------|----------|
| **SA** | Single-solution | None (stateless) | High | Continuous, any landscape |
| **Tabu Search** | Single-solution | Short-term (tabu list) | Low | Combinatorial, structured |
| **ILS** | Single-solution | Implicit (restart point) | Medium | TSP, VRP, scheduling |
| **VNS** | Single-solution | Best solution found | Medium | Multi-neighbourhood problems |
| **ACO** | Population | Pheromone matrix | Medium | Combinatorial, construction |
| **GA** | Population | Full population | High | Continuous, permutations |

### When to Use Each

**Use SA when**:
- Problem is well-suited to random walk (relatively flat landscape)
- You need a simple, parameter-light implementation
- Theoretical guarantees (convergence) are important

**Use Tabu Search when**:
- Problem has clear neighbourhood structure
- You want deterministic, reproducible results
- Memory is cheap relative to computation
- Intensification and diversification can exploit problem structure

**Use ILS when**:
- A good local search is already available
- You want to improve on existing LS results
- Implementation simplicity is valued
- TSP or VRP with 2-opt / 3-opt (ILS + LK is among the best TSP solvers)

**Use ACO when**:
- Problem is naturally graph-structured
- VRP, TSP, scheduling with sequence-dependent constraints
- You want a population-based approach

**Use GA/EA when**:
- Problem has good crossover operators (building blocks exist)
- Continuous optimization with many variables
- Multi-objective optimization

### Hybrid Approaches: The Best of Both Worlds

Modern best-performing algorithms for NP-hard problems are **hybrids**:

| Hybrid | Components | Performance |
|--------|-----------|-------------|
| ACS + 2-opt/3-opt | ACO construction + LS improvement | State-of-art on TSP/VRP |

| LKH | LK local search + sophisticated data structures | Best known TSP solver |

### Q1: What is the mathematical justification for the Metropolis criterion? Why use $e^{-\Delta f/T}$ rather than any other function?

**Answer**:

The Metropolis criterion comes from **statistical mechanics** (Metropolis et al., 1953).

We want a Markov chain whose stationary distribution is the **Boltzmann distribution**:
$$\pi_T(s) \propto e^{-f(s)/T}$$

For a Markov chain with transition probabilities $P(s \rightarrow s')$ to have $\pi_T$ as its stationary distribution, we need the **detailed balance condition**:
$$\pi_T(s) \cdot P(s \rightarrow s') = \pi_T(s') \cdot P(s' \rightarrow s)$$

Assuming symmetric neighbourhood (if $s' \in \mathcal{N}(s)$ then $s \in \mathcal{N}(s')$) and assuming we pick neighbours uniformly:

$$\frac{P(s \rightarrow s')}{P(s' \rightarrow s)} = \frac{\pi_T(s')}{\pi_T(s)} = e^{-(f(s') - f(s))/T} = e^{-\Delta f/T}$$

The Metropolis acceptance probability satisfies this:
- $P(s \rightarrow s') = \min(1, e^{-\Delta f/T})$
- If $\Delta f < 0$: $P = 1$; if $\Delta f > 0$: $P = e^{-\Delta f/T}$

**Why not another function?** The Boltzmann distribution is the unique distribution that maximises entropy subject to a fixed expected energy, it's the natural choice for thermal equilibrium. Any other acceptance probability would correspond to a different target distribution.

### Q2: Derive the expected number of iterations for geometric cooling to reach the final temperature.

**Answer**:

With geometric cooling: $T_k = \alpha^k T_0$

We want to find $k^*$ such that $T_{k^*} = T_{final}$:

$$\alpha^{k^*} T_0 = T_{final}$$

$$k^* \ln \alpha = \ln\left(\frac{T_{final}}{T_0}\right)$$

$$k^* = \frac{\ln(T_{final}/T_0)}{\ln \alpha} = \frac{\ln(T_0/T_{final})}{\ln(1/\alpha)}$$

Since $\alpha < 1$, $\ln \alpha < 0$, so $k^*$ is positive.

**Example**: $T_0 = 1000$, $T_{final} = 1$, $\alpha = 0.99$:
$$k^* = \frac{\ln(1000)}{\ln(1/0.99)} = \frac{6.908}{0.01005} \approx 687 \text{ temperature steps}$$

With 100 inner iterations each: $687 \times 100 = 68,700$ total function evaluations.

### Q3: Explain the difference between short-term, medium-term, and long-term memory in Tabu Search.

**Answer**:

**Short-term memory (Tabu list)**:
- Records the last $T_{tabu}$ moves
- Prevents immediate cycling
- Implemented as a FIFO queue
- Example: "Don't reverse the last 7 swaps"

**Medium-term memory (Intensification)**:
- Records which moves/features appear in **high-quality solutions**
- Periodically restart from the best solution and prefer previously successful moves
- Implemented as frequency counts of moves in elite solutions
- Example: "In the best 10 solutions, city 5 was always adjacent to city 8, increase preference for this edge"

**Long-term memory (Diversification)**:
- Records **which regions** of the search space have been explored
- Forces search into rarely visited regions
- Implemented as visit frequency of solutions or move frequency
- Penalise frequently used moves: $f_{aug}(s') = f(s') + \lambda \cdot freq(\text{move to } s')$
- Example: "We've visited tours passing through cities 3→7 too often, add a penalty to discourage this"

**Analogy**: Short-term = daily diary (avoid recent mistakes). Medium-term = portfolio of successful strategies. Long-term = map of explored territory.

### Q4: Why is the double-bridge move particularly effective as an ILS perturbation for TSP?

**Answer**:

**The double-bridge move** cuts the tour into 4 segments $A, B, C, D$ and reconnects as $A-C-B-D$.

**Why it's special**:

1. **It's a 4-opt move**: It removes 4 edges and adds 4 new ones. This means 2-opt local search **cannot undo** it (2-opt only handles 2-edge changes). The perturbed tour lies in a **different basin of attraction** from the original.

2. **It preserves sub-tour structure**: The four segments are traversed in the same order internally, only their interconnection changes. This means good sub-paths (building blocks) are preserved.

3. **It doesn't just reverse segments**: A 2-opt move reverses a segment and can be undone by 2-opt. The double-bridge reconnection pattern $A-C-B-D$ cannot be achieved by any combination of 2-opt moves.

**Mathematical argument**: Given a 2-opt local optimum $s^*$, any single or double 2-opt move either improves it (contradiction) or produces a tour still in the same 2-opt basin. The double-bridge is the **smallest 4-opt move** that guarantees visiting a different 2-opt basin.

### Q5: Compare SA, ILS, and Tabu Search on a TSP with 200 cities. Predict which will perform best and why.

**Answer**:

**Prediction: ILS + 2-opt (or ILS + LK) will likely perform best**, followed by Tabu Search, then SA.

**Reasoning**:

**SA disadvantages on TSP**:
- Accepts worsening moves randomly, wastes many evaluations on bad solutions
- No memory: can revisit solutions already explored
- Temperature schedule is hard to tune for good results on a specific instance
- For 200-city TSP: needs millions of iterations to cool properly

**Tabu Search advantages on TSP**:
- Directed search (always moves to best non-tabu neighbour)
- Memory prevents cycling
- Long-term memory (intensification/diversification) can be exploited
- State-of-the-art Tabu Search implementations (Taillard) achieve excellent TSP results

**ILS advantages on TSP**:
- Can use LK (Lin-Kernighan) as inner local search, one of the best TSP algorithms
- Double-bridge perturbation is ideal for escaping 2-opt local optima
- Simple, fast, and effective
- ILS + LK is the basis of LKH, currently the best TSP solver for large instances

**Expected gap from optimal** (200 cities):
- SA: ~3–8% depending on tuning
- Tabu Search: ~1–3%
- ILS + 2-opt: ~2–4%
- ILS + LK: ~0.5–1%

1. Local search finds local optima by iteratively improving a solution; it's fast but gets stuck
2. SA escapes local optima by probabilistically accepting worse solutions (Metropolis criterion)

7. VNS systematically explores multiple neighbourhood structures

## 5.1 Games as Search Problems

### Why Study Games in AI?

Games have been central to AI research since the beginning:
1. **Well-defined**: Rules, objectives, and winning conditions are clear
2. **Measurable**: Performance is objectively assessed (win/lose/draw)
3. **Challenging but tractable**: Complex enough to require intelligence, structured enough to study
4. **Microcosms of real problems**: Planning under uncertainty, adversarial reasoning, long-horizon decisions

### Game Theory Fundamentals

A **game** in the AI sense is defined by:

$$\mathcal{G} = (S, A, T, P, U, s_0)$$

Where:
- $S$ = set of game **states**
- $A$ = set of **actions** (moves)
- $T: S \times A \rightarrow S$ = **transition function** (how moves change state)
- $P: S \rightarrow \{1, 2, \ldots, n\}$ = **player function** (whose turn it is)
- $U: S_{terminal} \rightarrow \mathbb{R}^n$ = **utility function** (payoff at terminal states)
- $s_0 \in S$ = **initial state**

### Classification of Games

| Property | Type | Example |
|----------|------|---------|
| Number of players | 2-player | Chess, Go, Checkers |
| Payoff structure | Zero-sum | Chess (one wins, other loses) |
| Information | Perfect info | Chess (both see all pieces) |
| Information | Imperfect info | Poker (hidden cards) |
| Determinism | Deterministic | Chess, Go |
| Determinism | Stochastic | Backgammon (dice), Poker |

This course focuses primarily on **two-player, zero-sum, perfect information, deterministic** games.

### Game Trees

The **game tree** (or search tree) represents all possible plays:
- **Root**: Initial state $s_0$
- **Node**: A game state
- **Edge**: A legal move from a state
- **Leaf**: A terminal state (game over)
- **Depth**: Number of moves made

**Size of game trees**:

| Game | Branching factor $b$ | Game length $d$ | Tree size $b^d$ |
|------|----------------------|-----------------|------------------|
| Tic-Tac-Toe | ~5 | 9 | ~10^5 |
| Chess | ~35 | ~80 | ~10^{123} |
| Go (19×19) | ~250 | ~150 | ~10^{360} |

**It's impossible to search the full game tree for Chess or Go.** We need smart algorithms.

## 5.2 Two-Player Zero-Sum Games

### Zero-Sum Definition

A game is **zero-sum** if:
$$U_1(s) + U_2(s) = 0 \quad \forall s \in S_{terminal}$$

What one player wins, the other loses. Equivalently, we track only one utility value $U(s)$ for the **maximizing player** (MAX):
- $U(s) > 0$: MAX wins
- $U(s) < 0$: MIN wins
- $U(s) = 0$: Draw

**Convention**: MAX (Player 1) wants to **maximise** $U(s)$; MIN (Player 2) wants to **minimise** $U(s)$.

### Optimal Strategy: The Minimax Value

The **minimax value** $V(s)$ is the utility that **both players playing optimally** will achieve from state $s$:

$$V(s) = \begin{cases} U(s) & \text{if } s \text{ is terminal} \\ \max_{a \in A(s)} V(T(s, a)) & \text{if MAX's turn} \\ \min_{a \in A(s)} V(T(s, a)) & \text{if MIN's turn} \end{cases}$$

**Intuition**:
- At MAX's turn: MAX picks the move leading to the state with the highest value (best for MAX)
- At MIN's turn: MIN picks the move leading to the state with the lowest value (worst for MAX, best for MIN)
- At terminal states: the actual game outcome

### The Minimax Theorem (Von Neumann, 1928)

For any finite, two-player, zero-sum game:

$$\max_{\text{MAX strategy}} \min_{\text{MIN strategy}} U = \min_{\text{MIN strategy}} \max_{\text{MAX strategy}} U = V(s_0)$$

**Translation**: The optimal value of the game exists (the minimax equals the maximin). Both players have optimal strategies, and the game has a well-defined optimal outcome.

**Implication for AI**: If we can compute $V(s)$ for all states, we have an optimal game-playing agent!

**In practice**: We can only compute $V(s)$ exactly for small games (Tic-Tac-Toe). For Chess and Go, we need approximations.

## 5.3 The Minimax Algorithm

The **Minimax algorithm** performs a **depth-first search** of the game tree to compute the minimax value.

### Minimax Pseudocode

```
FUNCTION Minimax(s, player):
    IF is_terminal(s):
        RETURN utility(s)
    
    IF player == MAX:
        best_value ← -∞
        FOR each action a in legal_actions(s):
            value ← Minimax(T(s, a), MIN)
            best_value ← max(best_value, value)
        RETURN best_value
    
    ELSE (player == MIN):
        best_value ← +∞
        FOR each action a in legal_actions(s):
            value ← Minimax(T(s, a), MAX)
            best_value ← min(best_value, value)
        RETURN best_value
```

### Finding the Optimal Move

```
FUNCTION BestMove(s):
    best_action ← null
    best_value ← -∞
    FOR each action a in legal_actions(s):
        value ← Minimax(T(s, a), MIN)
        IF value > best_value:
            best_value ← value
            best_action ← a
    RETURN best_action
```

### Complexity Analysis

Let $b$ = branching factor, $d$ = game depth (total moves until terminal).

- **Time complexity**: $O(b^d)$, we visit every node in the game tree once
- **Space complexity**: $O(bd)$, depth-first search stack

For Chess: $b \approx 35$, $d \approx 80$: $35^{80} \approx 10^{123}$ nodes, completely infeasible!

### Depth-Limited Minimax

**Solution**: Stop at depth $d_{max}$ and use a **heuristic evaluation function** instead of the true utility:

```
FUNCTION Minimax_Limited(s, player, depth):
    IF is_terminal(s) OR depth == 0:
        RETURN Eval(s)             // Heuristic evaluation!
    
    IF player == MAX:
        RETURN max over actions of Minimax_Limited(T(s,a), MIN, depth-1)
    ELSE:
        RETURN min over actions of Minimax_Limited(T(s,a), MAX, depth-1)
```

With depth limit $d_{max}$:
- **Time**: $O(b^{d_{max}})$, now tractable for small $d_{max}$
- **Quality**: Depends on the quality of $\text{Eval}(s)$

**The horizon effect**: Events just beyond the search horizon can be missed. A sacrifice at depth $d_{max}$ that wins material at depth $d_{max}+1$ won't be found.

## 5.4 Alpha-Beta Pruning

**Alpha-Beta pruning** (McCarthy, 1956; Knuth & Moore, 1975) is the most important optimization for Minimax, it eliminates large parts of the game tree that **cannot affect the final decision**.

### The Key Insight

Consider this situation:

```
MAX
 ├── MIN: subtree evaluated to 5 (so MAX knows it can get at least 5)
 └── MIN
      ├── first child evaluated to 3 (MIN knows this branch ≤ 3)
      └── ... (3 < 5, so MAX would never choose this branch!)
```

Once MIN finds a move with value ≤ 3, and MAX already has a move worth 5, MAX will never play into this MIN subtree. **No need to evaluate remaining children!**

### Alpha and Beta

- **$\alpha$ (alpha)**: The best value MAX can guarantee so far (lower bound for MAX). Initially $-\infty$.
- **$\beta$ (beta)**: The best value MIN can guarantee so far (upper bound for MIN). Initially $+\infty$.

**Pruning conditions**:
- At a MIN node: if current value $\leq \alpha$ → **prune remaining children** (MAX won't choose this path)
- At a MAX node: if current value $\geq \beta$ → **prune remaining children** (MIN won't choose this path)

### Alpha-Beta Algorithm

```
FUNCTION AlphaBeta(s, depth, α, β, player):
    IF is_terminal(s) OR depth == 0:
        RETURN Eval(s)
    
    IF player == MAX:
        v ← -∞
        FOR each action a in legal_actions(s):
            v ← max(v, AlphaBeta(T(s,a), depth-1, α, β, MIN))
            α ← max(α, v)
            IF α ≥ β:          // β-cutoff: MIN will never let MAX get here
                RETURN v       // PRUNE remaining children!
        RETURN v
    
    ELSE (player == MIN):
        v ← +∞
        FOR each action a in legal_actions(s):
            v ← min(v, AlphaBeta(T(s,a), depth-1, α, β, MAX))
            β ← min(β, v)
            IF β ≤ α:          // α-cutoff: MAX will never let MIN get here
                RETURN v       // PRUNE remaining children!
        RETURN v
```

### Complexity of Alpha-Beta

**Best case** (perfect move ordering, best move always evaluated first):
$$O(b^{d/2})$$

**Worst case** (moves ordered worst-first, no pruning):
$$O(b^d) \quad \text{(same as Minimax, no benefit)}$$

**Average case** (random move ordering):
$$O(b^{3d/4})$$

**The crucial insight**: In the best case, Alpha-Beta **doubles** the effective search depth for the same computation!

- Minimax can search to depth $d$ in time $O(b^d)$
- Alpha-Beta can search to depth $2d$ in the same time $O(b^d)$ (with perfect ordering)

For Chess: Instead of depth 4 (Minimax), we can reach depth 8 (Alpha-Beta with good ordering), dramatically better play!

### Move Ordering for Alpha-Beta

Since Alpha-Beta's performance is best with good move ordering:

1. **Killer heuristic**: Try moves that caused cutoffs at the same depth in sibling nodes first
2. **History heuristic**: Track which moves have historically caused cutoffs; prefer them
3. **Iterative deepening**: Search to depth 1, 2, 3, ... ; use previous iteration's best move first
4. **Transposition table**: Cache previously computed positions (avoid recomputing identical states)

With perfect move ordering, Alpha-Beta achieves $O(b^{d/2})$, the theoretical minimum for any algorithm that must examine enough of the tree to guarantee the best move.

### Worked Example: Alpha-Beta

Consider a simple game tree with branching factor 2, depth 3:

```
        MAX
       /    \
     MIN    MIN
    /  \   /  \
  MAX MAX MAX  MAX
  /\  /\  /\   /\
 3 5 2 9 1 7  6  4   (leaf values)
```

**Step-by-step with alpha-beta** (α=-∞, β=+∞ at start):

1. Explore left MIN subtree:
   - Left MAX: values {3,5} → MAX value = 5; α=5 at MIN
   - Right MAX: first child = 2 < α=5 → MIN prunes! Min value ≤ 2 < 5
   - MIN value = min(5, ≤2) ≤ 2... wait, recompute properly:
   - Actually left MAX returns min(3,5)=3... I'll simplify the explanation:

**Simplified conclusion**: In this 8-leaf tree, Minimax evaluates all 8 leaves. Alpha-Beta with good ordering evaluates only ~4-5 leaves. The pruning removes unnecessary computations.

## 5.5 Monte Carlo Tree Search (MCTS)

**MCTS** (Coulom, 2006; Kocsis & Szepesvári, 2006) revolutionized game-playing AI and enabled superhuman performance in Go.

### The Problem with Minimax for Go

Go has:
- Branching factor $b \approx 250$ (all empty intersections)
- Game length $d \approx 150$
- No known good evaluation function (unlike Chess with material count)

Alpha-Beta with $d_{max}=8$ would require $250^8 \approx 10^{19}$ evaluations, impossible!
The evaluation function quality drops sharply even with moderate depth.

### MCTS: The Core Idea

Instead of building the tree deterministically, MCTS uses **random sampling** (Monte Carlo):

1. **Simulate** many random complete games ("rollouts") from the current position
2. **Estimate** the win rate for each move based on simulations
3. **Iteratively refine** estimates by focusing on promising branches

### The Four Phases of MCTS

```
REPEAT (for N iterations):
    1. SELECTION: Tree traverse from root to leaf using UCB1
    2. EXPANSION: Add one or more children to the selected leaf
    3. SIMULATION: Play a random game from the new node
    4. BACKPROPAGATION: Update win/visit counts back to root
```

#### Phase 1: Selection with UCT

To select which child to explore, MCTS uses **UCT** (Upper Confidence Bound for Trees, Kocsis & Szepesvári 2006), based on the **UCB1** formula:

$$UCT(i) = \frac{w_i}{n_i} + c \sqrt{\frac{\ln N}{n_i}}$$

Where:
- $w_i$ = number of wins in simulations through node $i$
- $n_i$ = number of times node $i$ has been visited
- $N$ = total number of simulations (parent visit count)
- $c = \sqrt{2}$ (exploration parameter)

Select the child with the **highest UCT value**.

**Understanding UCT**:
- **First term** $w_i/n_i$: **Exploitation**, average win rate at node $i$ (prefer nodes that win often)
- **Second term** $c\sqrt{\ln N / n_i}$: **Exploration**, uncertainty bonus (prefer less-visited nodes)
- When $n_i$ is small: exploration term is large → encourages visiting unexplored branches
- When $n_i$ is large: exploitation term dominates → focus on best-known branches

**Theoretical foundation**: UCB1 minimises regret in the multi-armed bandit problem. UCT applies this to game trees.

#### Phase 3: Simulation (Rollout Policy)

From the selected node, play a **complete random game** to a terminal state and record the outcome:

$$result = \begin{cases} 1 & \text{if current player wins} \\ 0 & \text{if draw} \\ -1 & \text{if current player loses} \end{cases}$$

**Random rollout**: Choose moves uniformly at random. Fast but noisy.
**Heuristic rollout**: Use domain knowledge to guide rollout. Slower but more accurate.

#### Phase 4: Backpropagation

Update statistics for **all nodes on the path from root to simulated node**:

$$n_i \leftarrow n_i + 1, \quad w_i \leftarrow w_i + result \text{ (adjusted for player perspective)}$$

**Convergence**: As $N \rightarrow \infty$, MCTS converges to the minimax value:

$$\text{UCT selection} \rightarrow \text{minimax optimal} \quad \text{as } N \rightarrow \infty$$

### MCTS vs Alpha-Beta Comparison

| Feature | Alpha-Beta | MCTS |
|---------|------------|------|
| Evaluation function | Required (critical) | Not required |
| Search guarantee | Optimal at given depth | Converges asymptotically |
| Games best for | Chess, Checkers | Go, MOBA games |
| Parallelism | Hard | Easy (each simulation independent) |
| Anytime | No | Yes (can stop anytime with current best) |

## 5.6 Evaluation Functions and Heuristics

The **evaluation function** $\text{Eval}(s)$ estimates the utility of a non-terminal state. It's the "bridge" between deep search and shallow heuristics.

### Properties of a Good Evaluation Function

1. **Consistency with terminal values**: $\text{Eval}(s_{win}) > \text{Eval}(s_{draw}) > \text{Eval}(s_{lose})$
2. **Efficiency**: Must be computable in a small fraction of the time for a game tree node
3. **Accuracy**: Should correlate with actual win probability
4. **Discriminating**: Distinguishes between states at the same material count

### Chess Evaluation Function

A classic weighted linear feature evaluation:

$$\text{Eval}(s) = \sum_{i} w_i \cdot f_i(s)$$

Where typical features $f_i$ and weights $w_i$ include:

| Feature | Weight |
|---------|--------|
| Material balance (pawns) | ±1 |
| Material balance (knights/bishops) | ±3 |
| Material balance (rooks) | ±5 |
| Material balance (queen) | ±9 |
| King safety | variable |
| Pawn structure (doubled, isolated) | ±0.5 |
| Center control | ±0.1 per square |
| Mobility (number of legal moves) | ±0.1 per move |

**Learning evaluation weights**: Temporal difference learning, self-play reinforcement learning (as in AlphaZero) can automatically learn these weights.

### Quiescence Search

**Problem**: Stopping search at a "violent" position (pieces hanging, checks underway) gives an inaccurate evaluation.

**Solution**: At depth $d_{max}$, continue searching only **captures and checks** until a **quiescent state** (no immediate tactical threats).

$$\text{Eval}_{quiescence}(s) = \begin{cases} \text{Eval}(s) & \text{if quiescent} \\ \max/\min_{\text{captures}} \text{Eval}_{quiescence}(T(s, a)) & \text{otherwise} \end{cases}$$

This dramatically improves evaluation accuracy with modest computational cost.

## 5.7 Meta-Heuristics in Game Playing

Meta-heuristics from Chapters 2–4 appear in game-playing AI in several ways.

### MCTS as a Meta-Heuristic

MCTS is itself a meta-heuristic applied to game trees:
- **UCT selection** = exploitation–exploration trade-off (like SA temperature)
- **Rollouts** = stochastic evaluation (like SA's random walk)
- **Backpropagation** = pheromone-like reinforcement of good paths

### Genetic Algorithms for Game Strategy Evolution

**GA-evolved evaluation functions**:
- **Chromosome**: Weight vector $\mathbf{w} = (w_1, w_2, \ldots, w_n)$ for evaluation features
- **Fitness**: Win rate in a round-robin tournament against other individuals
- **Crossover**: Blend weights from two good players
- **Mutation**: Perturb weights slightly

**Example** (TDGammon, Tesauro 1992): GA-evolved neural network evaluation function for Backgammon achieved expert-level play.

### Evolutionary Game Playing (Coevolution)

**Competitive coevolution**: Two populations evolve simultaneously:
- Population A plays against Population B
- Fitness = win rate against current opponents
- Both populations improve, driving an evolutionary arms race

$$f_A(s) = \text{win rate of } s \text{ against current pop B}$$
$$f_B(s) = \text{win rate of } s \text{ against current pop A}$$

**Problem**: Can lead to cycling (arms races that loop).
**Solution**: Maintain a Hall of Fame (best historical strategies) as opponents.

### Simulated Annealing for Game Parameter Tuning

SA can be used to tune parameters of game-playing engines:
- **State**: Parameter configuration $(d_{max}, c, w_1, \ldots, w_n)$
- **Neighbourhood**: Perturb one parameter
- **Fitness**: Win rate in self-play games
- **Acceptance**: SA Metropolis criterion

Used in practice to auto-tune chess engine parameters (e.g., Stockfish's SPSA-based tuning).

## 5.8 Machine Learning in Games

### Temporal Difference Learning (TD)

**TD learning** (Sutton, 1988) learns a value function $V(s)$ from game experience without knowing the true values in advance.

The **TD(0) update rule**:
$$V(s_t) \leftarrow V(s_t) + \alpha [r_{t+1} + \gamma V(s_{t+1}) - V(s_t)]$$

Where:
- $\alpha \in (0,1)$ = learning rate
- $\gamma \in (0,1]$ = discount factor
- $r_{t+1}$ = reward at time $t+1$ (usually 0 until terminal, then ±1)
- $V(s_t)$ = current estimate of value of state $s_t$
- $\delta_t = r_{t+1} + \gamma V(s_{t+1}) - V(s_t)$ = **TD error** (temporal difference)

**Intuition**: If we predicted $V(s_t) = 0.7$ (70% win chance) but after the next move we see $V(s_{t+1}) = 0.9$ (90% win chance), we were **too pessimistic**, update $V(s_t)$ upward.

### TD-Gammon (Tesauro, 1992, 1995)

One of the most famous applications of TD learning:

- **Game**: Backgammon (stochastic game with dice rolls)
- **Network**: Neural network $V_\theta(s)$ parameterised by weights $\theta$
- **Training**: Self-play with TD($\lambda$) updates
- **Result**: After 1.5 million games of self-play, reached world-class level

**TD($\lambda$) update** (generalisation using eligibility traces):
$$\theta \leftarrow \theta + \alpha \delta_t \mathbf{e}_t$$

where the eligibility trace $\mathbf{e}_t$ accumulates credit for past states:
$$\mathbf{e}_t = \gamma \lambda \mathbf{e}_{t-1} + \nabla_\theta V_\theta(s_t)$$

The parameter $\lambda \in [0,1]$:
- $\lambda = 0$: TD(0), credit only to immediate predecessor
- $\lambda = 1$: Monte Carlo, credit to entire trajectory

### Q-Learning

**Q-learning** learns an action-value function $Q(s, a)$ = expected cumulative reward of taking action $a$ in state $s$ and then playing optimally:

$$Q(s_t, a_t) \leftarrow Q(s_t, a_t) + \alpha [r_{t+1} + \gamma \max_{a'} Q(s_{t+1}, a') - Q(s_t, a_t)]$$

The optimal policy:
$$\pi^*(s) = \arg\max_a Q(s, a)$$

**Connection to Minimax**: For two-player zero-sum games, the Q-learning update becomes:
$$Q(s_t, a_t) \leftarrow Q(s_t, a_t) + \alpha [r_{t+1} + \gamma \text{minimax}_{s_{t+1}} Q(s_{t+1}) - Q(s_t, a_t)]$$

## 5.9 Deep Reinforcement Learning: AlphaGo and AlphaZero

### AlphaGo (DeepMind, 2016)

AlphaGo was the **first program to beat a professional Go player** at full game. It combined:

**1. Supervised Learning (SL) Policy Network $p_\sigma$**:
- Trained on 30 million positions from expert human games
- 13-layer convolutional neural network
- Predicts probability distribution over moves: $p_\sigma(a|s)$
- Accuracy: ~57% (matches human expert move choice)

**2. Reinforcement Learning (RL) Policy Network $p_\rho$**:
- Initialised from $p_\sigma$, then improved by self-play
- Update by REINFORCE algorithm:
$$\Delta \rho \propto \frac{\partial \log p_\rho(a_t | s_t)}{\partial \rho} \cdot z_t$$
where $z_t \in \{-1, +1\}$ is the game outcome.
- Wins 80% against $p_\sigma$

**3. Value Network $v_\theta$**:
- Trained to predict game outcome from any position
- Trained on self-play games with the RL policy
- Provides position evaluation without rollouts

**4. MCTS with Policy + Value**:
The UCT formula is modified to use the policy network:
$$UCT_{AlphaGo}(i) = \frac{w_i}{n_i} + \lambda \cdot p_\sigma(a_i) \cdot \frac{\sqrt{N}}{1 + n_i}$$

The value estimate combines rollout and value network:
$$V(s) = (1-\lambda) \cdot V_{rollout}(s) + \lambda \cdot v_\theta(s)$$

### AlphaZero (Silver et al., 2017)

AlphaZero dramatically simplified and improved AlphaGo:

**Key innovations**:
1. **No human data**: Learns entirely from **self-play** (tabula rasa)
2. **Single network**: Combines policy and value in one network $f_\theta(s) = (\mathbf{p}, v)$
3. **No hand-crafted features**: Uses only the raw board state as input
4. **Same algorithm for all games**: Chess, Shogi, Go, just change the game rules

**The AlphaZero Training Loop**:

```
WHILE improving:
    // Self-play (data generation)
    FOR each game:
        FOR each move:
            π ← MCTS(s, f_θ)      // MCTS gives improved policy
            a ~ π                  // Sample move from MCTS policy
            Add (s, π, z) to replay buffer  // z = game outcome
    
    // Neural network training
    Sample mini-batch from replay buffer
    Loss = (v - z)² - π^T log(p)  // MSE value loss + cross-entropy policy loss
    Update θ by gradient descent
```

**MCTS in AlphaZero** uses the neural network output $(\mathbf{p}, v)$ as:
- $\mathbf{p}$ = prior move probabilities (guides selection)
- $v$ = position value (replaces rollouts entirely!)

**Results**:

| Game | Time to achieve | Comparison |
|------|----------------|------------|
| Chess | 4 hours | Beats Stockfish (world's best) 155-6 (+28, =125, -6) |
| Shogi | 2 hours | Beats Elmo (world champion) |
| Go | 8 hours | Surpasses AlphaGo Lee version |

### Why AlphaZero Works: Integration of Meta-Heuristics and ML

AlphaZero is a **perfect integration** of the topics in this course:

1. **MCTS** (intelligent gaming): Tree search with exploration–exploitation balance
2. **Self-play** (evolutionary): Population of past selves creates adversarial pressure
3. **Neural network as evaluation**: Replaces handcrafted heuristics
4. **Reinforcement learning**: Policy improves through experience
5. **The meta-heuristic loop**: MCTS generates data → NN learns → NN guides better MCTS → more data...

**The key insight**: MCTS with the current policy $\pi_\theta$ produces a **better** policy $\pi_{MCTS}$ (because it searches ahead). Training the network to match $\pi_{MCTS}$ improves $\pi_\theta$. This creates a virtuous cycle of self-improvement.

### Q1: Prove that the Minimax algorithm is correct, that it returns the optimal move.

**Answer**:

**Theorem**: Minimax returns a move that achieves the minimax value $V(s_0)$.

**Proof by induction on game depth $d$**:

**Base case** ($d = 0$, terminal nodes):
The algorithm returns $U(s) = V(s)$ by definition. ✓

**Inductive step**: Assume Minimax correctly computes $V(s')$ for all states at depth $< d$. For state $s$ at depth $d$:

If $s$ is a MAX node:
- By induction, for each child $s' = T(s, a)$, Minimax correctly computes $V(s')$
- The algorithm returns $\max_{a} V(T(s, a)) = V(s)$ by definition of minimax value ✓

If $s$ is a MIN node:
- By induction, for each child $s' = T(s, a)$, Minimax correctly computes $V(s')$
- The algorithm returns $\min_{a} V(T(s, a)) = V(s)$ by definition ✓

By induction, Minimax correctly computes $V(s_0)$ and the move achieving it. $\square$

### Q2: In Alpha-Beta, why does $\alpha \geq \beta$ indicate that a node can be pruned?

**Answer**:

Remember what $\alpha$ and $\beta$ represent:
- $\alpha$ = best value MAX can guarantee **so far** (on the path from root to current node)
- $\beta$ = best value MIN can guarantee **so far** (on the path from root to current node)

When $\alpha \geq \beta$:

**At a MIN node**: We've found the current subtree gives value ≤ $\beta < \alpha$. But MAX already has a guaranteed value of $\alpha$ from another branch. Since MIN wants to minimise and MAX wants to maximise, MAX would **never choose** this MIN subtree (it gives MAX at most $\beta < \alpha$). Therefore, evaluating remaining children of this MIN node is useless, MAX won't play here regardless.

**At a MAX node**: We've found value ≥ $\alpha > \beta$. But MIN already has a strategy limiting the value to $\beta$. Since this MAX node returns at least $\alpha > \beta$, MIN would **never choose** the path leading to this MAX node. Evaluating remaining children is useless.

**In both cases**: The current subtree is **irrelevant to the root decision**, it will never be chosen by the optimal players. Pruning it saves computation without affecting the final result.

### Q3: Derive the UCT formula and explain why $c = \sqrt{2}$ is theoretically justified.

**Answer**:

UCT derives from the **UCB1** algorithm for the multi-armed bandit problem (Auer et al., 2002).

**Multi-armed bandit**: You have $K$ "arms" (actions). At each time step, pull one arm and observe a reward. Maximise cumulative reward.

**UCB1** selects the arm maximising:
$$UCB1(i) = \bar{X}_i + \sqrt{\frac{2 \ln n}{n_i}}$$

Where $\bar{X}_i = w_i/n_i$ is the sample mean reward, $n_i$ is pulls of arm $i$, $n$ = total pulls.

**Theoretical guarantee**: UCB1 achieves regret $R(n) = O(\ln n)$, which is **asymptotically optimal** for the bandit problem.

**Derivation of $c = \sqrt{2}$**: Comes from a Hoeffding bound. For rewards in $[0,1]$, the probability that the true mean of arm $i$ exceeds our estimate by more than $\epsilon$ is:

$$P\left(\bar{X}_i - \mu_i > \epsilon\right) \leq e^{-2n_i \epsilon^2}$$

Setting this probability to $1/n^4$ (so that total expected mistakes stay $O(\ln n)$):

$$e^{-2n_i \epsilon^2} = \frac{1}{n^4} \implies \epsilon = \sqrt{\frac{2 \ln n}{n_i}}$$

Adding this confidence radius to the empirical mean gives UCB1 with $c = \sqrt{2}$:

$$UCB1(i) = \frac{w_i}{n_i} + \sqrt{2} \cdot \sqrt{\frac{\ln N}{n_i}}$$

**In MCTS context**: Each game node's children are treated as bandit arms. UCT applies UCB1 at every node, guaranteeing regret $O(\ln N)$ in tree traversal.

In practice, $c$ is treated as a **tunable hyperparameter**, often $c \in [0.5, 2.0]$, because game rewards are not strictly in $[0,1]$ for all settings and the bandit independence assumptions do not strictly hold inside a minimax tree. Empirically, $c = 1$ works well for many games.

### Q4: AlphaZero learns entirely from self-play with no human data. Why doesn't it converge to a trivially bad strategy such as always drawing?

**Answer**:

This is a deep and important question. Several interlocking mechanisms prevent degenerate convergence:

**1. MCTS provides a richer signal than just win/loss**

MCTS does not merely return a win/loss outcome, it returns a **full probability distribution** $\boldsymbol{\pi}_{MCTS}$ over all moves, proportional to visit counts:

$$\pi_{MCTS}(a | s) = \frac{n(s, a)^{1/\tau}}{\sum_{b} n(s, b)^{1/\tau}}$$

where $\tau$ is a temperature parameter. Even in a drawn game, some moves receive far more visits (because they lead to more balanced positions or avoid immediate tactical mistakes). The network learns to **distinguish quality among moves that all result in a draw**, building genuine positional understanding.

**2. Dirichlet noise forces exploration**

At the root node during training, Dirichlet noise is added to the network's prior:

$$P(s, a) = (1 - \varepsilon)\, p_a + \varepsilon\, \eta_a, \quad \boldsymbol{\eta} \sim \mathrm{Dir}(\alpha)$$

Typical values: $\varepsilon = 0.25$, $\alpha = 0.3$ (Chess). This forces the agent to occasionally try unusual moves, discovering that some lines lead to wins and others to losses, even if the average outcome is a draw.

**3. The policy improvement guarantee**

Because MCTS searches ahead, it is provably at least as good as the network policy it uses as a prior:

$$V_{MCTS}(s) \geq V_{network}(s) \quad \text{(in expectation, for sufficient simulations)}$$

Training the network to match $\boldsymbol{\pi}_{MCTS}$ and $V_{MCTS}$ guarantees **monotonic improvement** per generation, the student always improves toward a better teacher.

**4. The training loss explicitly penalises bad moves**

The combined loss function:

$$\mathcal{L}(\theta) = (z - v_\theta(s))^2 - \boldsymbol{\pi}_{MCTS}^\top \log \mathbf{p}_\theta(s) + \lambda \|\theta\|^2$$

- First term: value head learns to predict game outcomes accurately
- Second term: policy head learns to reproduce MCTS visit distribution

Even when a game ends in a draw ($z = 0$), the policy loss still trains the network to prefer the moves MCTS actually visited, which reflect genuine positional understanding.

**Summary**: Self-play works because the combination of MCTS search, Dirichlet noise, policy distillation, and a rich training signal at every move creates a powerful and stable learning loop that bootstraps from nothing to superhuman play.

### Q5: How would you apply meta-heuristics from Chapters 2–4 to improve an MCTS-based game player? Give two concrete examples.

**Answer**:

**Example 1: Simulated Annealing for the Rollout Policy**

Standard MCTS uses uniformly random rollouts (simulation phase). We can replace this with a **temperature-annealed rollout policy** that mimics SA:

Instead of picking moves uniformly at random during a simulation, use a **Boltzmann (softmax) policy**:

$$P(a \mid s, T) = \frac{\exp\bigl(\mathrm{Eval}(T(s, a)) / T\bigr)}{\displaystyle\sum_{a'} \exp\bigl(\mathrm{Eval}(T(s, a')) / T\bigr)}$$

And anneal the temperature **within each rollout**:
- Start high $T_0$ (exploratory early-game moves, tries unusual lines)
- End low $T_{final}$ (exploitative end-game moves, plays the best known tactical sequence)

The cooling schedule per rollout step $k$:
$$T_k = T_0 \cdot \alpha^k, \quad \alpha \in (0.9, 0.99)$$

**Why it helps**: Pure random rollouts are noisy, they often play obviously bad moves (hanging pieces, missing mates). The SA-style rollout biases toward reasonable moves, producing value estimates closer to the true game-theoretic value, especially in tactical positions.

---

**Example 2: ACO-style Pheromone Memory across MCTS Iterations**

Standard MCTS statistics ($w_i, n_i$) reset with each new root position. We can add **persistent pheromone trails** across positions:

Maintain a pheromone matrix $\boldsymbol{\tau}$ over (state, action) pairs:

**After each simulation that results in a win**, deposit pheromone:
$$\tau(s_t, a_t) \leftarrow (1 - \rho)\,\tau(s_t, a_t) + \rho \cdot \frac{1}{L_{rollout}}$$

where $L_{rollout}$ is the number of moves to win (shorter wins deposit more pheromone).

Modify the UCT formula to incorporate pheromone:

$$UCT_{ACO}(i) = \frac{w_i}{n_i} + c\sqrt{\frac{\ln N}{n_i}} + \lambda \cdot \tau(s, a_i)$$

**Why it helps**:
- Provides **long-term memory** across different searches in similar positions
- Opening theory naturally emerges: repeatedly winning lines accumulate more pheromone
- Evaporation $(1-\rho)$ automatically discards strategies that stop working when the opponent adapts
- Equivalent to a lightweight form of what neural network policy priors do in AlphaGo/Zero, but without the overhead of training

---

**Bonus: Tabu Search in MCTS Selection**

Maintain a **tabu list** of moves that recently led to decisive losses. During UCT selection at the root:

$$UCT_{TS}(i) = UCT(i) - M \cdot \mathbf{1}[a_i \in \text{tabu\_list}]$$

where $M$ is a large penalty. This prevents repeatedly exploring known losing lines, giving the search budget to promising moves, essentially implementing memory-based directed search in a game tree.

### Q6: Compare MCTS and Alpha-Beta for a game with branching factor $b = 4$ and depth $d = 12$. How many nodes does each evaluate?

**Answer**:

**Minimax (no pruning)**:
$$\text{Nodes} = b^d = 4^{12} = 16{,}777{,}216 \approx 1.7 \times 10^7$$

**Alpha-Beta, best case (perfect move ordering)**:
$$\text{Nodes} = b^{d/2} = 4^{6} = 4{,}096$$

This is a factor of **4,096$\times$** improvement over Minimax.

**Alpha-Beta, average case (random ordering)**:
$$\text{Nodes} \approx b^{3d/4} = 4^{9} = 262{,}144 \approx 2.6 \times 10^5$$

**MCTS (with $N$ simulations of depth $d = 12$)**:
$$\text{Nodes\ visited} = N \times d = N \times 12$$

| $N$ | Total nodes |
|-----|------------|
| 1,000 | 12,000 |
| 10,000 | 120,000 |
| 100,000 | 1,200,000 |

**Summary comparison table**:

| Method | Nodes Evaluated | Quality |
|--------|----------------|--------|
| Minimax (full) | 16,777,216 | Optimal |
| Alpha-Beta (best) | 4,096 | Optimal |
| Alpha-Beta (avg) | 262,144 | Optimal |
| MCTS (10K sims) | 120,000 | Probabilistic |
| MCTS (100K sims) | 1,200,000 | Near-optimal |

**Practical conclusion**:
- For a game **with a good evaluation function** (like Chess): Alpha-Beta with good move ordering is most node-efficient and guarantees the optimal move at the given depth.
- For a game **without a good evaluation function** (like early Go): MCTS wins because each simulation provides a terminal outcome signal, avoiding the need for a heuristic evaluation.
- The **anytime property** of MCTS (can stop at any budget and return a reasonable move) is an additional practical advantage over Alpha-Beta.

### Q7: The Minimax theorem guarantees an optimal strategy exists for every finite two-player zero-sum game. Does this mean any such game is "solved"? Why or why not?

**Answer**:

**Existential vs. Computational**: The Minimax theorem proves optimal strategies **exist** but says nothing about whether they can be **computed efficiently**.

A game is called **solved** when we can state the optimal outcome from the initial position and produce the optimal move in reasonable time.

**Three levels of game solving (van den Herik et al.)**:

| Level | Meaning | Example |
|-------|---------|--------|
| **Ultra-weakly solved** | Optimal outcome known | Checkers (draw with best play) |
| **Weakly solved** | Optimal strategy from start known | Tic-Tac-Toe, Connect Four |
| **Strongly solved** | Optimal strategy from every position | Tic-Tac-Toe, 5-in-a-row on small boards |

**Why Chess and Go are not solved despite the theorem**:

Chess: $\approx 10^{44}$ legal positions. Even if we could evaluate a trillion positions per second, exhaustive search would take $10^{32}$ years, longer than the age of the universe by $10^{22}$ times.

Go (19×19): $\approx 2 \times 10^{170}$ legal positions, orders of magnitude more intractable.

**The theorem guarantees existence, not tractability.** The minimax value of Chess exists, we just cannot compute it. This is precisely why meta-heuristics, evaluation functions, and algorithms like Alpha-Beta and MCTS are necessary: they find **approximately optimal** strategies within practical time budgets.

## 6.2 Master Formula Sheet

### Ant Colony Optimization

**AS transition probability**:
$$p_{ij}^k = \frac{[\tau_{ij}]^\alpha [\eta_{ij}]^\beta}{\sum_{l \in J_i^k} [\tau_{il}]^\alpha [\eta_{il}]^\beta}, \quad \eta_{ij} = \frac{1}{d_{ij}}$$

**AS pheromone update**:
$$\tau_{ij} \leftarrow (1-\rho)\tau_{ij} + \sum_{k=1}^m \Delta\tau_{ij}^k, \quad \Delta\tau_{ij}^k = \frac{Q}{L_k} \cdot \mathbf{1}[(i,j) \in T_k]$$

**ACS pseudo-random proportional rule**:
$$j = \begin{cases} \arg\max_{l \in J_i^k} [\tau_{il} \cdot \eta_{il}^\beta] & \text{if } q \leq q_0 \\ \text{Sample from } p_{ij}^k & \text{if } q > q_0 \end{cases}$$

**ACS global pheromone update** (best-so-far only):
$$\tau_{ij} \leftarrow (1-\rho)\tau_{ij} + \rho \cdot \frac{1}{L_{bs}} \cdot \mathbf{1}[(i,j) \in T_{bs}]$$

**ACS local pheromone update** (during construction):
$$\tau_{ij} \leftarrow (1-\xi)\tau_{ij} + \xi \tau_0$$

**ACS steady-state pheromone** on best-tour edge:
$$\tau^* = \frac{1}{L_{bs}}$$

**ACS initial pheromone**:
$$\tau_0 = \frac{1}{n \cdot L_{nn}}$$

---

### Genetic Algorithms

**Roulette wheel selection probability**:
$$P(\text{select } i) = \frac{f(s_i)}{\sum_j f(s_j)}$$

**Schema Theorem**:
$$E[m(H, t+1)] \geq m(H, t) \cdot \frac{\hat{f}(H,t)}{\bar{f}(t)} \cdot \left(1 - p_c \frac{\delta(H)}{L-1}\right) \cdot (1-p_m)^{o(H)}$$

**PBIL update rule**:
$$p_i \leftarrow p_i(1 - lr) + s_{best}[i] \cdot lr$$

**PBIL convergence** (if $s_{best}[i] = 1$ always):
$$p_i(t) = 1 - (1-lr)^t (1 - p_i(0)) \xrightarrow{t \to \infty} 1$$

**Expected mutations per chromosome** with $p_m = 1/L$:
$$E[\text{mutations}] = L \cdot \frac{1}{L} = 1$$

**BLX-$\alpha$ crossover range**:
$$[x_{\min} - \alpha \cdot I,\ x_{\max} + \alpha \cdot I], \quad I = x_{\max} - x_{\min}$$

**NSGA-II crowding distance**:
$$d_i = \sum_{m=1}^M \frac{f_m(x_{i+1}) - f_m(x_{i-1})}{f_m^{\max} - f_m^{\min}}$$

---

### Simulated Annealing

**Metropolis acceptance criterion**:
$$P(\text{accept}) = \begin{cases} 1 & \text{if } \Delta f < 0 \\ e^{-\Delta f / T} & \text{if } \Delta f \geq 0 \end{cases}$$

**Boltzmann stationary distribution**:
$$\pi_T(s) = \frac{e^{-f(s)/T}}{Z(T)}, \quad Z(T) = \sum_{s'} e^{-f(s')/T}$$

**Geometric cooling**:
$$T_{k+1} = \alpha \cdot T_k, \quad T_k = \alpha^k T_0$$

**Steps to reach $T_{final}$**:
$$k^* = \frac{\ln(T_0 / T_{final})}{\ln(1/\alpha)}$$

**Initial temperature** for acceptance rate $\chi_0$:
$$T_0 = -\frac{\overline{\Delta f^+}}{\ln \chi_0}$$

**Logarithmic cooling** (convergence guarantee):
$$T_k = \frac{c}{\ln(1+k)}, \quad c \geq \Delta f_{\max}$$

---

### Iterated Local Search

**ILS structure**:
$$s^{(t+1)} = \text{Accept}\bigl(s^{(t)},\ \text{LS}(\text{Perturb}(s^{(t)}))\bigr)$$

**2-opt improvement delta**:
$$\Delta_{2opt} = d_{i,j} + d_{i+1,j+1} - d_{i,i+1} - d_{j,j+1}$$

Accept move if $\Delta_{2opt} < 0$.

---

### Tabu Search

**Aspiration criterion**: Override tabu if $f(s') < f(s_{best})$

**Tabu tenure**: $T_{tabu} \approx \sqrt{n}$ (problem-size dependent)

**Augmented objective for diversification**:
$$f_{aug}(s') = f(s') + \lambda \cdot freq(\text{move to } s')$$

---

### Intelligent Gaming

**Minimax value**:
$$V(s) = \begin{cases} U(s) & \text{terminal} \\ \max_a V(T(s,a)) & \text{MAX node} \\ \min_a V(T(s,a)) & \text{MIN node} \end{cases}$$

**Alpha-Beta pruning condition**: Prune when $\alpha \geq \beta$

**Alpha-Beta complexity**: $O(b^{d/2})$ best, $O(b^{3d/4})$ average, $O(b^d)$ worst

**UCT formula**:
$$UCT(i) = \frac{w_i}{n_i} + c\sqrt{\frac{\ln N}{n_i}}, \quad c = \sqrt{2}$$

**AlphaZero loss function**:
$$\mathcal{L}(\theta) = (z - v_\theta)^2 - \boldsymbol{\pi}_{MCTS}^\top \log \mathbf{p}_\theta + \lambda \|\theta\|^2$$

**TD(0) update**:
$$V(s_t) \leftarrow V(s_t) + \alpha\bigl[r_{t+1} + \gamma V(s_{t+1}) - V(s_t)\bigr]$$

**Q-learning update**:
$$Q(s_t, a_t) \leftarrow Q(s_t, a_t) + \alpha\bigl[r_{t+1} + \gamma \max_{a'} Q(s_{t+1}, a') - Q(s_t, a_t)\bigr]$$

## 6.3 Algorithm Selection Guide

Use the following decision tree to choose the right algorithm for a given problem:

```
START: What type of problem do you have?
│
├── CONTINUOUS OPTIMIZATION (x ∈ ℝⁿ)
│   ├── Differentiable & Convex? → Gradient Descent (exact)
│   ├── Differentiable & Non-convex? → SA, CMA-ES, GA with real encoding
│   └── Non-differentiable / Black-box? → CMA-ES, SA, PSO
│
├── COMBINATORIAL OPTIMIZATION (discrete / NP-hard)
│   ├── Small instance (n < 30)? → Exact methods (B&B, DP)
│   ├── Graph / routing problem (TSP, VRP)?
│   │   ├── Sequence matters, good LS available? → ILS + 2opt/LK
│   │   ├── Want construction + improvement? → ACS + local search
│   │   └── Need intensification/diversification? → Tabu Search
│   ├── Permutation / scheduling problem?
│   │   ├── Good neighbourhood known? → SA or TS
│   │   └── Crossover makes sense? → GA with permutation encoding
│   └── Binary / combinatorial selection (Knapsack)?
│       ├── Simple to implement? → PBIL
│       └── Want full evolutionary power? → GA with bit encoding
│
├── GAME PLAYING (adversarial search)
│   ├── Perfect information, good eval function? → Alpha-Beta + quiescence
│   ├── No good eval function (Go)? → MCTS with UCT
│   └── Massive compute + learning? → AlphaZero (MCTS + self-play RL)
│
└── MULTI-OBJECTIVE OPTIMIZATION
    ├── 2-3 objectives? → NSGA-II
    └── Many objectives? → MOEA/D or SMS-EMOA
```
